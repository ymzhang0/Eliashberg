# Hubbard-Stratonovich Transformation Derivation



## 1. Standard Gaussian Identity

We start with the auxiliary Gaussian integral for a variable $y$:


\begin{equation}

    \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}y^2\right) dy = \sqrt{2\pi}

\end{equation}





## 2. Shift Transformation

Introduce a shift $y \to y - cx^2$ to couple $y$ with $x^2$. The integration measure remains unchanged ($dy$):


\begin{equation}

\begin{aligned}
    \sqrt{2\pi} &= \int_{-\infty}^{\infty} \exp\left[-\frac{1}{2}(y - cx^2)^2\right] dy \\
    &= \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}y^2 + cyx^2 - \frac{1}{2}c^2 x^4\right) dy \\
    &= \exp\left(-\frac{1}{2}c^2 x^4\right) \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}y^2 + cyx^2\right) dy
\end{aligned}

\end{equation}



Rearranging to isolate the quartic term:


\begin{equation}

    \exp\left(\frac{1}{2}c^2 x^4\right) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}y^2 + cyx^2\right) dy

\end{equation}





## 3. Coefficient Matching

To match the target term $\exp(bx^4)$, we set $\frac{1}{2}c^2 = b$, which implies $c = \sqrt{2b}$. Substituting this back:


\begin{equation}

    \exp(bx^4) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}y^2 + \sqrt{2b}\,x^2 y\right) dy

\end{equation}





## 4. Full Integral Transformation

Substitute the above identity into the original integral $I = \int_{-\infty}^{\infty} \exp(-ax^2 + bx^4) dx$:


\begin{equation}

\begin{aligned}
    I &= \int_{-\infty}^{\infty} e^{-ax^2} \left[ \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} e^{-\frac{1}{2}y^2 + \sqrt{2b}\,x^2 y} dy \right] dx \\
      &= \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} dy \, e^{-\frac{1}{2}y^2} \underbrace{\int_{-\infty}^{\infty} dx \, \exp\left[ -\left(a - y\sqrt{2b}\right)x^2 \right]}_{\text{Gaussian integral over } x}
\end{aligned}

\end{equation}




## 5. Integrating out $x$

Performing the inner Gaussian integral over $x$:


\begin{equation}

    I = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \exp\left(-\frac{1}{2}y^2\right) \sqrt{\frac{\pi}{a - y\sqrt{2b}}} \, dy

\end{equation}


We can absorb the $\frac{1}{\sqrt{2\pi}}$ into the Gaussian integral over $y$:

\begin{equation}

    I = \int_{-\infty}^{\infty} \frac{\exp\left(-y^2\right)}{\sqrt{{a - 2\sqrt{b}y}}} \, dy

\end{equation}

## 6. Saddle Point Approximation for the Integral with a Denominator

We aim to approximate the following integral:

$$
I = \int_{-\infty}^{\infty} dy \, \frac{e^{-y^2}}{\sqrt{a - 2\sqrt{b}y}}
$$

## 7. Effective Action Formulation

To perform a systematic expansion, we absorb the pre-factor (the denominator) into the exponent. We rewrite the integrand as $e^{-\Phi(y)}$, where $\Phi(y)$ is the total effective action:

$$
\Phi(y) = y^2 + \frac{1}{2} \ln(a - 2\sqrt{b}y)
$$




## 8. Saddle Point Condition

The saddle point $y_0$ is defined by the stationary condition of the action $\Phi'(y_0) = 0$.

## 9. Derivation
$$
\frac{d\Phi}{dy} = 2y + \frac{1}{2} \frac{-2\sqrt{b}}{a - 2\sqrt{b}y} = 2y - \frac{\sqrt{b}}{a - 2\sqrt{b}y} = 0
$$

Rearranging this yields a quadratic equation:
$$
2y(a - 2\sqrt{b}y) = \sqrt{b} \implies 4\sqrt{b}y^2 - 2ay + \sqrt{b} = 0
$$

## 10. Solution

Solving for $y$, and selecting the branch that remains finite (and close to 0) as $b \to 0$ (the perturbative limit):

$$
y_0 = \frac{a - \sqrt{a^2 - 4b}}{4\sqrt{b}}
$$



## 11. Gaussian Fluctuations (Second Derivative)

We expand $\Phi(y)$ to the second order around $y_0$:
$$
\Phi(y) \approx \Phi(y_0) + \frac{1}{2} \Phi''(y_0) (y - y_0)^2
$$

## 12. Calculating $\Phi''$
$$
\Phi''(y) = \frac{d}{dy} \left( 2y - \frac{\sqrt{b}}{a - 2\sqrt{b}y} \right) = 2 - \frac{2b}{(a - 2\sqrt{b}y)^2}
$$

## 13. Simplification at $y_0$
Using the saddle point condition $\frac{\sqrt{b}}{a - 2\sqrt{b}y_0} = 2y_0$, we substitute the fraction term:

$$
\Phi''(y_0) = 2 - 2 \left( \frac{\sqrt{b}}{a - 2\sqrt{b}y_0} \right)^2 = 2 - 2(2y_0)^2 = 2(1 - 4y_0^2)
$$


## 14. Final Approximation Result

Integrating the Gaussian fluctuations:
$$
I \approx e^{-\Phi(y_0)} \int_{-\infty}^{\infty} dy \, e^{-\frac{1}{2} \Phi''(y_0) (y - y_0)^2} = e^{-\Phi(y_0)} \sqrt{\frac{2\pi}{\Phi''(y_0)}}
$$

Substituting the explicit forms of $e^{-\Phi}$ and $\Phi''$:

1.  **Exponential Part:** $e^{-\Phi(y_0)} = \frac{e^{-y_0^2}}{\sqrt{a - 2\sqrt{b}y_0}}$
2.  **Fluctuation Part:** $\sqrt{\frac{2\pi}{2(1 - 4y_0^2)}} = \sqrt{\frac{\pi}{1 - 4y_0^2}}$

**Final Formula:**
$$
I \approx \frac{e^{-y_0^2}}{\sqrt{a - 2\sqrt{b}y_0}} \sqrt{\frac{\pi}{1 - 4y_0^2}}
$$

*Validity Condition:* The approximation holds strictly when the effective curvature is positive, i.e., $1 - 4y_0^2 > 0$, or $y_0 < 0.5$.


```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define Parameters ---
# Large 'a' and small 'b' to ensure the approximation holds
a = 10.0
b = 1.0

# --- 2. Define Domain and Functions ---
# Calculate the singularity position where the log argument becomes zero
y_singularity = a / (2 * np.sqrt(b))

# Define y-axis range, stopping just before the singularity
y = np.linspace(-3, y_singularity - 0.05, 500)

# A. Full quartic potential
def phi_quartic(x, a, b):
    return -a*x**2 + b*x**4

# B. Full Effective Action
def phi_exact(y, a, b):
    # Use np.maximum to prevent log(negative) errors
    safe_arg = np.maximum(a - 2 * np.sqrt(b) * y, 1e-9)
    # FIX: Use ** for power, not ^
    return y**2 + 0.5 * np.log(safe_arg)

# C. Unperturbed Action (Gaussian)
def phi_original(y):
    return y**2

# --- 3. Calculate Saddle Point Properties ---
# Theoretical saddle point position y0
y_0 = (a - np.sqrt(a**2 - 4*b)) / (4 * np.sqrt(b))

# Action value at the saddle point Phi(y0)
phi_0 = phi_exact(y_0, a, b)

# Second derivative (Curvature) at the saddle point: Phi''(y0) = 2(1 - 4*y0^2)
phi_pp_0 = 2 * (1 - 4 * y_0**2)

# C. Quadratic Approximation
# Phi_approx(y) = Phi(y0) + 0.5 * Phi''(y0) * (y - y0)^2
phi_approx_vals = phi_0 + 0.5 * phi_pp_0 * (y - y_0)**2

# --- 4. Plotting ---
plt.figure(figsize=(10, 6))

# Plot the three curves
plt.plot(y, phi_quartic(y, a, b), label=r'Quartic Potential $\Phi(y)$', color='blue', linewidth=2)
plt.plot(y, phi_exact(y, a, b), label=r'Full Action $\Phi(y)$', color='blue', linewidth=2)
plt.plot(y, phi_approx_vals, label=r'Quadratic Approx around $y_0$', color='red', linestyle='--', linewidth=2)
plt.plot(y, phi_original(y), label=r'Original Gaussian $y^2$', color='gray', linestyle=':', alpha=0.7)

# Mark the Saddle Point and Original Minimum
plt.scatter([y_0], [phi_0], color='red', zorder=5, s=80, label='Saddle Point $y_0$')
plt.scatter([0], [0], color='gray', marker='x', zorder=5, s=80, label='Original Minimum ($y=0$)')

# Add a vertical line for the singularity
plt.axvline(x=y_singularity, color='black', linestyle='-.', alpha=0.3, label='Singularity')

# Formatting the plot
plt.title(f'Comparison of Full Action vs Quadratic Approximation\n($a={a}, b={b}$)')
plt.xlabel('$y$')
plt.ylabel('Action value $\Phi(y)$')

# Set limits to focus on the relevant potential well
plt.ylim(-10, np.max(phi_exact(y[y<2.5], a, b)) + 1)
plt.xlim(-3, y_singularity + 0.5)

plt.legend(loc='best', frameon=True, shadow=True)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

Remark:

1. The exact effective action match perfectly with the truncated quadratic action.
2. The effective action loses the two minima of the original quartic potential.


## Free Energy Calculation via Saddle Point Approximation


In this section, we move beyond the partition function $Z$ (the integral value) and calculate the **Free Energy** $F$. This allows us to physically analyze how the interaction term (parameter $b$) affects the stability and entropy of the system.

### 1. Definition

The partition function $Z$ in the saddle point approximation is given by:

$$
Z \approx e^{-\Phi(y_0)} \sqrt{\frac{2\pi}{\Phi''(y_0)}}
$$

The dimensionless Free Energy is defined as:

$$
F = - \ln Z
$$

Substituting the approximation into the definition:

$$
F \approx - \ln \left( e^{-\Phi(y_0)} \sqrt{\frac{2\pi}{\Phi''(y_0)}} \right) = \Phi(y_0) - \frac{1}{2} \ln \left( \frac{2\pi}{\Phi''(y_0)} \right)
$$



### 2. Decomposition: Mean Field vs. Fluctuations

It is insightful to decompose the Free Energy into two distinct physical contributions:

$$
F \approx F_{\text{MF}} + F_{\text{fluc}}
$$

The first term corresponds to the value of the effective action at the saddle point. This represents the energy of the classical ground state configuration.

$$
F_{\text{MF}} = \Phi(y_0) = y_0^2 + \frac{1}{2} \ln(a - 2\sqrt{b}y_0)
$$

The second term comes from the Gaussian integral around the saddle point. It relates to the **Entropy** of the fluctuations.

$$
F_{\text{fluc}} = \frac{1}{2} \ln \Phi''(y_0) - \frac{1}{2} \ln(2\pi)
$$

Using our derived expression for the curvature $\Phi''(y_0) = 2(1 - 4y_0^2)$:

$$
F_{\text{fluc}} \approx \frac{1}{2} \ln \left[ 2(1 - 4y_0^2) \right] + \text{const}
$$

* **Physical Insight:** Since $y_0$ increases with interaction $b$, the term $(1-4y_0^2)$ becomes smaller.
* **Curvature $\downarrow$**: The potential well becomes "softer" (wider).
* **Fluctuations $\uparrow$**: The particle has more phase space to explore.
* **Entropy $\uparrow$**: Higher entropy leads to a lower (more negative) Free Energy contribution.



### 3. Numerical Analysis

The following Python code visualizes how the interaction parameter $b$ affects the total free energy and its components.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Parameters ---
a = 10.0
# We vary b from 0 (Gaussian) to a value where approximation holds
# Constraint: y0 < 0.5 for stability (curvature > 0)
# Theoretical max b roughly where 4*y0^2 approaches 1.
b_vals = np.linspace(0.001, 5.0, 100) 

# --- 2. Calculation Functions ---

def calculate_y0(a, b):
    # Saddle point formula derived previously
    # Using the minus branch for continuity with b=0
    return (a - np.sqrt(a**2 - 4*b)) / (4 * np.sqrt(b))

def get_free_energy_components(a, b_array):
    F_total = []
    F_MF = []
    F_fluc = []
    
    for b in b_array:
        y0 = calculate_y0(a, b)
        
        # 1. Effective Action at saddle point (Mean Field)
        # phi = y^2 + 0.5 * ln(a - 2*sqrt(b)*y)
        phi_0 = y0**2 + 0.5 * np.log(a - 2 * np.sqrt(b) * y0)
        
        # 2. Curvature (Fluctuations)
        # phi'' = 2(1 - 4y0^2)
        curvature = 2 * (1 - 4 * y0**2)
        
        # Free Energy contributions
        f_mf_val = phi_0
        # F_fluc = 0.5 * ln(curvature) - 0.5 * ln(2pi)
        f_fluc_val = 0.5 * np.log(curvature) - 0.5 * np.log(2 * np.pi)
        
        F_MF.append(f_mf_val)
        F_fluc.append(f_fluc_val)
        F_total.append(f_mf_val + f_fluc_val)
        
    return np.array(F_total), np.array(F_MF), np.array(F_fluc)

# --- 3. Compute ---
F_tot, F_mf, F_fluc = get_free_energy_components(a, b_vals)

# Normalize relative to b -> 0 case to see the shift clearly
F_tot_norm = F_tot - F_tot[0]
F_mf_norm = F_mf - F_mf[0]
F_fluc_norm = F_fluc - F_fluc[0]

# --- 4. Plotting ---
plt.figure(figsize=(10, 6))

plt.plot(b_vals, F_tot_norm, label=r'Total Free Energy Change $\Delta F$', color='black', linewidth=2.5)
plt.plot(b_vals, F_mf_norm, label=r'Mean Field Part $\Delta \Phi(y_0)$', color='blue', linestyle='--')
plt.plot(b_vals, F_fluc_norm, label=r'Fluctuation Part $\frac{1}{2}\ln \Phi^{\prime\prime}$', color='red', linestyle='-.')

plt.title(f'Effect of Interaction $b$ on Free Energy ($a={a}$)')
plt.xlabel('Interaction Parameter $b$')
plt.ylabel('Change in Free Energy')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Annotations
plt.arrow(1.0, -0.05, 0, -0.1, head_width=0.1, head_length=0.02, fc='k', ec='k')
plt.text(1.2, -0.1, 'System becomes "softer"\n(Entropy increases)', fontsize=10)

plt.show()
```

# Complex Scalar Field: A Toy Model for Channel Selection

To understand why channel selection (Direct vs. Cooper) is necessary even without the complications of Grassmann algebra (fermions), we study a simple **Complex Scalar Field** $z$. This model demonstrates that the pairing structure arises purely from the particle-antiparticle ($z$ vs $z^*$) degree of freedom.

## 1. The Model

Consider a partition function for a complex variable $z = x + iy$:

$$
Z = \int \frac{dz d\bar{z}}{2\pi i} e^{-S[z, \bar{z}]}
$$

The action includes a mass term and a quartic interaction:
$$
S = \epsilon |z|^2 + g |z|^4
$$
where $|z|^4 = z^* z z^* z$.

## 2. Channel 1: Density / Direct Channel

We group the interaction term as the square of the density ($n = z^* z$):
$$
|z|^4 = (z^* z) \cdot (z^* z)
$$

### 2.1 HS Transformation
We introduce a **real** auxiliary field $\phi$ (analogous to the Hartree potential) to decouple the square term:
$$
e^{-g (z^* z)^2} \propto \int d\phi \, \exp\left( -\frac{\phi^2}{4g} - i\phi (z^* z) \right)
$$

### 2.2 Effective Action
Substituting this back, the action becomes quadratic in $z$:
$$
S_{\text{eff}}[\phi, z] = \frac{\phi^2}{4g} + z^* (\epsilon + i\phi) z
$$

Integrating out the complex field $z$ using the standard Gaussian formula $\int d^2z e^{-A |z|^2} \propto A^{-1} = e^{-\ln A}$:

$$
S_{\text{eff}}^{\text{Direct}}[\phi] = \frac{\phi^2}{4g} + \ln(\epsilon + i\phi)
$$

* **Structure**: The fluctuation term is a **scalar logarithm**.
* **Physics**: Describes density fluctuations.




### Expansion in the Direct Channel: Random Phase Approximation (RPA)



In this section, we derive the Random Phase Approximation (RPA) effective action for a complex scalar field by expanding the logarithm term in the Direct (Density) Channel.

### 1. The Effective Action

We start with the effective action derived from the Hubbard-Stratonovich transformation in the density channel:

$$
S_{\text{eff}}[\phi] = \frac{\phi^2}{4g} + \ln(\epsilon + i\phi)
$$

Here, $\phi$ is the real auxiliary field coupled to the density $z^* z$, $\epsilon$ is the mass/energy gap, and $g$ is the interaction strength.

### 2. Taylor Expansion (The RPA Logic)

The RPA corresponds to expanding the action to the second order in the auxiliary field fluctuations (Gaussian approximation).

#### 2.1 Logarithm Expansion
We expand $\ln(\epsilon + i\phi)$ assuming small fluctuations $|i\phi| \ll \epsilon$:

$$
\ln(\epsilon + i\phi) = \ln \left[ \epsilon \left( 1 + \frac{i\phi}{\epsilon} \right) \right] = \ln \epsilon + \ln \left( 1 + \frac{i\phi}{\epsilon} \right)
$$

Using the Taylor series $\ln(1+x) \approx x - \frac{x^2}{2}$:

$$
\ln \left( 1 + \frac{i\phi}{\epsilon} \right) \approx \frac{i\phi}{\epsilon} - \frac{1}{2} \left( \frac{i\phi}{\epsilon} \right)^2
$$
$$
= \frac{i}{\epsilon}\phi - \frac{1}{2} \frac{-\phi^2}{\epsilon^2} = \frac{i}{\epsilon}\phi + \frac{\phi^2}{2\epsilon^2}
$$

*(Note: The quadratic term is positive because $i^2 = -1$.)*

#### 2.2 Resulting Action
Substituting back into $S_{\text{eff}}$ (and ignoring the constant $\ln \epsilon$):

$$
S_{\text{RPA}}[\phi] \approx \underbrace{\frac{i}{\epsilon}\phi}_{\text{Hartree Term}} + \frac{\phi^2}{4g} + \frac{\phi^2}{2\epsilon^2}
$$

### 3. Physical Interpretation

We can regroup the quadratic terms to define the **RPA Susceptibility** (or inverse propagator):

$$
S_{\text{RPA}}[\phi] = i \frac{\phi}{\epsilon} + \frac{1}{2} \phi^2 \left( \frac{1}{2g} + \Pi \right)
$$

### Key Quantities:
1.  **Polarization Bubble ($\Pi$):**
    $$\Pi = \frac{1}{\epsilon^2}$$
    This represents the response of the non-interacting Bose gas to the field. It comes from the curvature of the $\ln$ potential.

2.  **Screened Interaction ($g_{\text{eff}}$):**
    The effective interaction is the inverse of the quadratic coefficient (propagator):
    $$D_{\text{RPA}} \sim \left( \frac{1}{2g} + \Pi \right)^{-1} = \frac{2g}{1 + 2g \Pi}$$
    The denominator $(1 + 2g\Pi)$ is the **Dielectric Function**. It shows that the bare interaction $g$ is screened (weakened) by the medium's polarization.

3.  **Hartree Term ($i\phi/\epsilon$):**
    The linear term represents the mean-field energy shift. In a full calculation, we would shift $\phi \to \phi - i\langle \phi \rangle$ to absorb this term into the ground state energy.

### 4. Numerical Comparison

The following code compares the exact logarithmic potential with the quadratic RPA approximation.



```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
g = 1.0        
epsilon = 1.0  

# --- Functions ---
# We visualize the real part of the potential by setting phi = -iV (real potential V)
# S_exact(V) = ln(epsilon + V)
# S_rpa(V) = ln(epsilon) + V/epsilon - V^2/(2*epsilon^2)

def potential_exact(V, eps):
    return np.log(eps + V)

def potential_rpa(V, eps):
    return np.log(eps) + V/eps - (V**2)/(2 * eps**2)

# --- Plotting ---
V_vals = np.linspace(-0.8, 0.8, 100)
S_ex = potential_exact(V_vals, epsilon)
S_rpa = potential_rpa(V_vals, epsilon)

plt.figure(figsize=(8, 5))

plt.plot(V_vals, S_ex, label=r'Exact Loop Term: $\ln(\epsilon + V)$', color='blue', linewidth=2)
plt.plot(V_vals, S_rpa, label=r'RPA Approx: Linear + Quadratic', color='red', linestyle='--', linewidth=2)

plt.scatter([0], [np.log(epsilon)], color='black', zorder=5)
plt.text(0.05, np.log(epsilon)+0.1, "Expansion Point", fontsize=10)

plt.title(f'RPA Approximation vs Exact Potential\n($\epsilon={epsilon}$)')
plt.xlabel('Field Fluctuation $V$')
plt.ylabel('Effective Potential')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## 3. Channel 2: Cooper / Pairing Channel



We group the interaction term to pair particles with particles (and holes with holes):
$$
|z|^4 = (z^* z^*) \cdot (z z)
$$

### 3.1 HS Transformation
We introduce a **complex** auxiliary field $\Delta$ (analogous to the Superconducting gap):
$$
e^{-g (z^* z^*) (z z)} \propto \int d\Delta d\Delta^* \, \exp\left( -\frac{|\Delta|^2}{g} + \Delta^* (z z) + \Delta (z^* z^*) \right)
$$

### 3.2 Nambu Spinor Formulation
The action now contains anomalous terms like $zz$ and $z^*z^*$:
$$
S_{\text{quad}} = \epsilon z^* z - \Delta^* z z - \Delta z^* z^*
$$
To write this in matrix form, we must treat $z$ and $z^*$ as a vector. We introduce the **Nambu Spinor** $\Psi = \begin{pmatrix} z \\ z^* \end{pmatrix}$. Note that for bosons, symmetrization requires care, but the structure is:

$$
S_{\text{quad}} = \frac{1}{2} \begin{pmatrix} z^* & z \end{pmatrix} 
\underbrace{\begin{pmatrix} \epsilon & -2\Delta \\ -2\Delta^* & \epsilon \end{pmatrix}}_{\mathbb{M}} 
\begin{pmatrix} z \\ z^* \end{pmatrix}
$$

### 3.3 Effective Action
Integrating out the field $\Psi$ involves the determinant of the matrix $\mathbb{M}$. For complex bosons, the integral yields $(\det \mathbb{M})^{-1/2}$, leading to a $+\frac{1}{2} \ln \det$ term (Fermions would have a minus sign).

$$
\det \mathbb{M} = \epsilon^2 - 4|\Delta|^2
$$

$$
S_{\text{eff}}^{\text{Cooper}}[\Delta] = \frac{|\Delta|^2}{g} + \frac{1}{2} \ln(\epsilon^2 - 4|\Delta|^2)
$$

* **Structure**: The fluctuation term comes from a **Matrix Determinant**.
* **Physics**: Describes pairing instability. The singularity at $\epsilon^2 = 4|\Delta|^2$ indicates the boundary of stability for the bosonic system.

---

## 4. Note on Integration Measure

A crucial point in the derivation is the definition of the integration measure. Although we write $dz d\bar{z}$, we are integrating over a **single complex plane** (2 real degrees of freedom).

The transformation from Cartesian coordinates $(x, y)$ is given by the Jacobian:
$$
z = x + iy, \quad \bar{z} = x - iy
$$
$$
dz \wedge d\bar{z} = (dx + i dy) \wedge (dx - i dy) = -2i (dx \wedge dy)
$$

Thus, the measure transforms as:
$$
\int \frac{dz d\bar{z}}{2\pi i} (\dots) = \int \frac{-2i dx dy}{2\pi i} (\dots) = \int \frac{dx dy}{\pi} (\dots)
$$
This confirms that $z$ and $\bar{z}$ are not independent variables but coupled coordinates of the same 2D physical space.

## Cooper Channel Expansion with Matsubara Frequencies

We perform an RPA-like expansion of the effective action in the Cooper channel for a complex scalar field, incorporating finite temperature effects via Matsubara summation.

### 1. RPA Expansion (Static Limit)

The effective action derived from the toy model is:
$$
S_{\text{eff}}[\Delta] = \frac{|\Delta|^2}{g} + \frac{1}{2} \ln(\epsilon^2 - 4|\Delta|^2)
$$

Expanding the logarithm for small order parameter $|\Delta|$:
$$
\ln(\epsilon^2 - 4|\Delta|^2) = \ln \epsilon^2 + \ln\left( 1 - \frac{4|\Delta|^2}{\epsilon^2} \right) \approx \text{const} - \frac{4|\Delta|^2}{\epsilon^2}
$$

Substituting this back gives the quadratic (RPA) action:
$$
S_{\text{RPA}} = |\Delta|^2 \left( \frac{1}{g} - \frac{2}{\epsilon^2} \right)
$$
The term $-2/\epsilon^2$ represents the **screening** (or renormalization) of the bare stiffness $1/g$ by the fluctuations of the scalar field.

### 2. Finite Temperature Formulation

To capture the temperature dependence $\Delta(T)$, we replace the static logarithm with a sum over Matsubara frequencies. For a complex scalar field (bosons), $\omega_n = 2n\pi T$.

$$
\frac{1}{2} \ln(\dots) \longrightarrow \frac{1}{2} T \sum_{n=-\infty}^{\infty} \ln \left[ (\omega_n^2 + \epsilon^2) - 4|\Delta|^2 \right]
$$

#### 2.1 The Pair Susceptibility
Expanding the term inside the sum to order $|\Delta|^2$:
$$
S_{\text{RPA}}(T) = |\Delta|^2 \left( \frac{1}{g} - \chi_{\text{pair}}(T) \right)
$$

The temperature-dependent pair susceptibility is:
$$
\chi_{\text{pair}}(T) = - \frac{1}{2} T \sum_n \left( - \frac{4}{\omega_n^2 + \epsilon^2} \right) = 2 T \sum_n \frac{1}{\omega_n^2 + \epsilon^2}
$$

#### 2.2 Performing the Sum (Bosons)
Using the standard summation identity for bosons:
$$
T \sum_n \frac{1}{\omega_n^2 + \epsilon^2} = \frac{1}{2\epsilon} \coth\left( \frac{\epsilon}{2T} \right)
$$

Thus:
$$
\chi_{\text{pair}}(T) = \frac{1}{\epsilon} \coth\left( \frac{\epsilon}{2T} \right)
$$

### 3. The Instability Condition (Gap Equation)

The system becomes unstable (forming a condensate $\Delta \neq 0$) when the effective stiffness vanishes:
$$
\frac{1}{g} - \frac{1}{\epsilon} \coth\left( \frac{\epsilon}{2T_c} \right) = 0
$$

### Comparison: Boson vs. Fermion
* **Bosons (This Model):** $\chi \propto \coth(\epsilon/2T)$. As $T \to 0$, $\chi \to 1/\epsilon$. As $T \to \infty$, $\chi \to \infty$. The instability is driven by high temperature (thermal population).
* **Fermions (BCS Theory):** $\chi \propto \frac{1}{\epsilon}\tanh(\epsilon/2T)$. (Or $\frac{1}{\xi}\tanh(\xi/2T)$ integrated). As $T \to 0$, $\chi$ is maximized. The instability is driven by low temperature (Fermi surface nesting).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Parameters
g = 0.8; epsilon = 1.0

# Functions
def r_coeff(T, g, eps):
    # Effective stiffness r(T)
    # Using coth(x) = 1/tanh(x)
    if T < 1e-4: return 1.0/g - 1.0/eps
    val = eps / (2*T)
    chi = (1.0/eps) * (1.0 / np.tanh(val))
    return 1.0/g - chi

T_vals = np.linspace(0.1, 5.0, 100)
r_vals = [r_coeff(t, g, epsilon) for t in T_vals]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(T_vals, r_vals, label=r'$r(T)$', linewidth=2)
plt.axhline(0, color='gray', linestyle='--')
plt.title(r'Instability in Bosonic Cooper Channel ($S \sim \Delta^\dagger \Delta$)')
plt.xlabel('Temperature $T$')
plt.ylabel(r'Coefficient $1/g - \chi(T)$')
plt.legend()
plt.grid(True, alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
epsilon = 1.0
g = 0.5 

# --- Effective Actions ---

def S_direct(V, eps, g):
    # Density Channel: Scalar Log
    # Using V = i*phi (Hartree potential)
    # S = -V^2/4g + ln(eps + V)
    valid = (eps + V) > 0
    val = np.full_like(V, np.nan)
    val[valid] = -(V[valid]**2)/(4*g) + np.log(eps + V[valid])
    return val

def S_cooper(Delta, eps, g):
    # Cooper Channel: Matrix Log (Determinant)
    # S = |Delta|^2/g + 0.5 * ln(eps^2 - 4|Delta|^2)
    valid = (eps**2 - 4*Delta**2) > 0
    val = np.full_like(Delta, np.nan)
    val[valid] = (Delta[valid]**2)/g + 0.5 * np.log(eps**2 - 4*Delta[valid]**2)
    return val


# --- Plotting ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Direct Channel
V_vals = np.linspace(-0.9, 2.0, 200)
ax1.plot(V_vals, S_direct(V_vals, epsilon, g), 'b-', linewidth=2)
ax1.set_title(r'Direct Channel: $S \sim \ln(\epsilon + V)$')
ax1.set_xlabel('Potential $V$')
ax1.grid(True)

# Cooper Channel
epsilon = 1.0
g = 0.5 

D_vals = np.linspace(-epsilon/2, epsilon/2, 200)
ax2.plot(D_vals, S_cooper(D_vals, epsilon, g), 'r-', linewidth=2, label='g=0.5, epsilon=1.0')

epsilon = 1.0
g = 0.2

D_vals = np.linspace(-epsilon/2, epsilon/2, 200)
ax2.plot(D_vals, S_cooper(D_vals, epsilon, g), 'r-', linewidth=2, label='g=0.1, epsilon=1.0')

ax2.set_title(r'Cooper Channel: $S \sim \frac{1}{2}\ln(\epsilon^2 - 4\Delta^2)$')
ax2.set_xlabel('Gap Parameter $\Delta$')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

# Interaction Expansion for a Two-State Fermion System

## 1. Model Setup


A more realistic example is that the field $\phi$ is a function of momentum $p$. 

Consider the interaction part of the action:

$$
S_{\text{int}} = \frac{C}{2} \sum_{p, p', q} \bar{\psi}_{p+q} \bar{\psi}_{p'-q} V(\mathbf{q}) \psi_{p'} \psi_p
$$

Where $C = T^3/L^3$ is the coefficient. For simplicity, we restrict the momentum space to two discrete states: $p, p' \in \{p_1, p_2\}$.

**Key Properties of Grassmann Numbers:**
* **Anti-commutativity:** $\psi_a \psi_b = - \psi_b \psi_a$
* **Nilpotency:** $\psi_a \psi_a = 0$ (Pauli exclusion principle)

---


## 2. Expansion of the Summation



Due to $\psi_p \psi_p = 0$, terms vanish when $p = p'$. We only need to sum over $p \neq p'$.
There are two symmetric cases: $(p=p_1, p'=p_2)$ and $(p=p_2, p'=p_1)$. We expand the first case in detail.

To ensure the output momentum indices remain within the set $\{p_1, p_2\}$, the momentum transfer $q$ can only take two specific values:

### Case A ($q=0$): Direct / Hartree Term
The output indices remain unchanged ($p_1 \to p_1, p_2 \to p_2$):
$$
\text{Term}_1 = \bar{\psi}_{p_1} \bar{\psi}_{p_2} V(0) \psi_{p_2} \psi_{p_1}
$$

### Case B ($q = p_2 - p_1$): Exchange / Fock Term
The output indices are swapped ($p_1 \to p_2, p_2 \to p_1$):
$$
\text{Term}_2 = \bar{\psi}_{p_2} \bar{\psi}_{p_1} V(p_2 - p_1) \psi_{p_2} \psi_{p_1}
$$

Using the anti-commutation relation $\bar{\psi}_{p_2} \bar{\psi}_{p_1} = - \bar{\psi}_{p_1} \bar{\psi}_{p_2}$, we rewrite the exchange term as:
$$
\text{Term}_2 = - V(p_2 - p_1) \bar{\psi}_{p_1} \bar{\psi}_{p_2} \psi_{p_2} \psi_{p_1}
$$

---


## 3. Resulting Effective Interaction


Summing all contributions (including the symmetric $p \leftrightarrow p'$ case, which cancels the factor $1/2$), the total interaction action becomes:

$$
S_{\text{int}}^{\text{2-state}} = C \underbrace{\left[ V(0) - V(p_2 - p_1) \right]}_{g_{\text{eff}}} \bar{\psi}_{p_1} \bar{\psi}_{p_2} \psi_{p_2} \psi_{p_1}
$$

---


## 4. Hubbard-Stratonovich Transformation

The four-fermion interaction term can be decoupled in the Density, Exchange and Cooper channels. These alternative schemes reveal different physical instabilities, such as Charge Density Waves (CDW) or Ferromagnetism or Superconductivity.

### Direct / Hartree Channel

We use **Fierz-like rearrangements** to rewrite Eq.12 before applying the Hubbard-Stratonovich (HS) transformation.

This channel focuses on the interaction between particle densities $\hat{n}_p = \bar{\psi}_p \psi_p$.

#### Operator Rearrangement
We commute $\psi_{p_2}$ and $\psi_{p_1}$ (generating a minus sign) to group operators by their momentum index:

$$
S_{\text{int}} = - C g_{\text{eff}} (\bar{\psi}_{p_1} \psi_{p_1}) (\bar{\psi}_{p_2} \psi_{p_2})
$$

Using the algebraic identity for densities $n_1 n_2 \approx \frac{1}{2}(n_1 + n_2)^2$ (ignoring $n^2$ terms which only shift the chemical potential), we approximate the interaction as a square of the total density:

$$
S_{\text{int}}^{\text{dens}} \approx - \frac{C g_{\text{eff}}}{2} \left( \bar{\psi}_{p_1} \psi_{p_1} + \bar{\psi}_{p_2} \psi_{p_2} \right)^2
$$

#### HS Transformation
We introduce a real auxiliary scalar field $\phi(\tau)$ to decouple the quadratic term.
* **Field Interpretation:** $\phi$ acts as a fluctuating electrostatic potential (Hartree field).

$$
\exp\left( -S_{\text{int}}^{\text{dens}} \right) = \int \mathcal{D}\phi \exp \left[ - \int d\tau \left( \frac{\phi^2}{2 C g_{\text{eff}}} + i\phi (\bar{\psi}_{p_1} \psi_{p_1} + \bar{\psi}_{p_2} \psi_{p_2}) \right) \right]
$$

#### Fermionic Path Integral
The total fermionic action becomes diagonal in momentum space (no mixing between $p_1$ and $p_2$):

$$
S_{\text{fermion}} = \sum_{\omega_n} \left[ \bar{\psi}_{p_1} (-i\omega_n + \xi_{p_1} + i\phi) \psi_{p_1} + \bar{\psi}_{p_2} (-i\omega_n + \xi_{p_2} + i\phi) \psi_{p_2} \right]
$$

Since the matrix is diagonal, the determinant is simply the product of the integrals for each state.

#### Effective Action
Integrating out $\psi_{p_1}$ and $\psi_{p_2}$:

$$
S_{\text{eff}}[\phi] = \frac{\beta \phi^2}{2 C g_{\text{eff}}} - \sum_n \ln(-i\omega_n + \xi_{p_1} + i\phi) - \sum_n \ln(-i\omega_n + \xi_{p_2} + i\phi)
$$

---

#### Evaluating the Matsubara Sum

We aim to simplify the effective action term involving the summation over fermionic Matsubara frequencies $\omega_n = (2n+1)\pi/\beta$.

$$
\Sigma = \sum_{n=-\infty}^{\infty} \ln \left( -i\omega_n + \epsilon \right)
$$

Here, $\epsilon = \xi_p + i\phi$ represents the effective single-particle energy shifted by the auxiliary field.

Using standard contour integration techniques or the infinite product representation of the partition function, the following identity holds for fermions:

$$
\text{Tr} \ln (-i\omega_n + \hat{H}) = \ln \det (1 + e^{-\beta \hat{H}}) + \text{const}
$$

Specifically for a scalar eigenvalue $\epsilon$:
$$
\sum_n \ln(-i\omega_n + \epsilon) = \ln(1 + e^{-\beta \epsilon}) + \frac{\beta \epsilon}{2} + C
$$
*(Note: The $\beta \epsilon / 2$ term is a regularization factor often absorbed into the chemical potential or ignored if we only care about differences).*

Substituting this back into your expression for $S_{\text{eff}}$:

$$
S_{\text{eff}}[\phi] = \frac{\beta \phi^2}{2 C g_{\text{eff}}} - \ln\left(1 + e^{-\beta(\xi_{p_1} + i\phi)}\right) - \ln\left(1 + e^{-\beta(\xi_{p_2} + i\phi)}\right)
$$

The term $-\frac{1}{\beta} \ln(1 + e^{-\beta E})$ is exactly the **Grand Potential (Free Energy)** of a single fermion state with energy $E$.

Thus, the effective action describes:
1.  **Energy Cost:** The Gaussian term $\phi^2$ (energy stored in the auxiliary field).
2.  **Fermionic Feedback:** The sum of free energies of particles $p_1$ and $p_2$ moving in the fluctuating potential $i\phi$.

#### Expansion of the Effective Action (RPA Level)

We consider the fermionic term in the effective action for a single state $p$ (suppressing index):
$$
S_{\text{fermion}}[\phi] = - \sum_{n} \ln(-i\omega_n + \xi + i\phi)
$$

Assuming the auxiliary field $\phi$ (fluctuation) is small, we perform a Taylor expansion around $\phi = 0$.

** 1.1 Taylor Series **
$$
\ln(A + x) \approx \ln(A) + \frac{x}{A} - \frac{x^2}{2A^2}
$$
Here, $A = -i\omega_n + \xi$ and $x = i\phi$.

** First Order Term (Density) **
$$
S^{(1)} = - \sum_n \frac{i\phi}{-i\omega_n + \xi} = -i\phi \underbrace{ \sum_n \frac{1}{-i\omega_n + \xi} }_{\beta n_F(\xi)}
$$
This term couples $\phi$ to the average particle density $\langle n \rangle = n_F(\xi)$.

** Second Order Term (Polarization) **
$$
S^{(2)} = - \sum_n \left( - \frac{(i\phi)^2}{2(-i\omega_n + \xi)^2} \right) = \frac{\phi^2}{2} \sum_n \frac{1}{(-i\omega_n + \xi)^2}
$$
To evaluate the sum $\Pi = \sum_n \frac{1}{(-i\omega_n + \xi)^2}$, we note that:
$$
\frac{d}{d\xi} \left( \sum_n \frac{1}{-i\omega_n + \xi} \right) = \sum_n \frac{-1}{(-i\omega_n + \xi)^2} = -\Pi
$$
Since $\sum \frac{1}{-i\omega_n+\xi} = \beta n_F(\xi)$, we have:
$$
\Pi = -\frac{d}{d\xi} [\beta n_F(\xi)] = -\beta n_F'(\xi) = \beta n_F(\xi)(1 - n_F(\xi))
$$
This quantity $\Pi$ is the **static Lindhard susceptibility** (or compressibility) of the state.

Combining with the original $\phi^2$ term from the HS transformation:

$$
S_{\text{eff}}^{\text{RPA}}[\phi] \approx S_0 + \underbrace{(-i\phi \rho_{0})}_{\text{Source}} + \frac{1}{2} \phi^2 \left( \frac{\beta}{C g_{\text{eff}}} + \Pi \right)
$$

This describes a harmonic oscillator with a "renormalized" spring constant (dielectric function).

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Parameters ---
beta = 5.0
Cg = -1.0  # Repulsive interaction
xi = 0.5   # Single state energy

# --- 2. Functions ---
# A. Exact Action (Wick rotated i*phi -> V)
def action_exact(V, beta, Cg, xi):
    # Gaussian term: phi^2/2g -> -V^2/2g
    term_gaussian = - (V**2) / (2 * Cg) * beta
    # Log term
    term_fermion = - np.log(1 + np.exp(-beta * (xi + V)))
    return term_gaussian + term_fermion

# B. Quadratic Expansion (RPA) around V=0
def action_rpa(V, beta, Cg, xi):
    # 0th order: Free energy at V=0
    n_F = 1 / (1 + np.exp(beta * xi))
    S_0 = -np.log(1 + np.exp(-beta * xi))
    
    # 1st order coeff: - beta * density (Note: dS/dV = n)
    # Since S_fermion = -ln(1+e^(-beta(xi+V))), dS/dV = beta * e/(1+e) = beta * n_F
    coeff_1 = beta * n_F
    
    # 2nd order coeff (Susceptibility): beta^2 * (n^2 - n) ... wait, let's derive carefully
    # d^2S/dV^2 = d/dV (beta * n_F(xi+V)) = beta * dn/dV = beta * (-beta n(1-n))
    # Curvature from Fermions = -beta^2 * n_F * (1 - n_F)
    # Curvature from Gaussian = -beta / Cg
    
    curvature = -beta/Cg - (beta**2 * n_F * (1 - n_F))
    
    # Expansion: S(0) + S'(0)V + 0.5*S''(0)V^2
    return S_0 + coeff_1 * V + 0.5 * curvature * V**2

# --- 3. Plotting ---
V_vals = np.linspace(-1.5, 1.5, 200)
S_exact = action_exact(V_vals, beta, Cg, xi)
S_approx = action_rpa(V_vals, beta, Cg, xi)

plt.figure(figsize=(8, 5))
plt.plot(V_vals, S_exact, label='Full Action (Exact)', color='purple', linewidth=2)
plt.plot(V_vals, S_approx, label='Quadratic Expansion (RPA)', color='orange', linestyle='--', linewidth=2)

plt.scatter([0], [action_exact(0, beta, Cg, xi)], color='black', zorder=5, label='Expansion Point (V=0)')

plt.title(f'Density Channel: Exact vs RPA Expansion\n($\\beta={beta}, \\xi={xi}, Cg={Cg}$)')
plt.xlabel('Hartree Potential $V$')
plt.ylabel('Effective Action')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## RPA Free Energy: The Role of the Linear Term

In this section, we derive the Free Energy in the Random Phase Approximation (RPA). We explicitly address the presence of the linear term in the effective action and how it leads to the Hartree energy shift.

### Why is there a linear term?

In the single-variable saddle point approximation, we expanded around $y_0$ such that $S'(y_0)=0$, eliminating the linear term.
However, the standard RPA action is obtained by expanding around the **non-interacting ground state** ($\phi=0$).

$$
S_{\text{eff}}[\phi] \approx S_0 \underbrace{- i\rho_0 \phi}_{\text{Force}} + \frac{1}{2} A \phi^2
$$

* $\phi=0$ is **not** the saddle point of the interacting system due to the background charge density $\rho_0$.
* The linear term represents the "force" driving the system towards the true mean-field value.

### Derivation of Free Energy

To find the partition function $Z$, we integrate over the auxiliary field $\phi$. We must **complete the square** to handle the linear term.

### Completing the Square
We define the stiffness $A = \frac{\beta}{C g_{\text{eff}}} + \Pi$. The stationary point $\phi^*$ is found by:
$$
\frac{\partial S}{\partial \phi} = -i\rho_0 + A\phi = 0 \implies \phi^* = i \frac{\rho_0}{A}
$$

Substituting $\phi = \phi^* + \delta\phi$ into the action:
$$
S(\phi) = S_0 + \left( -i\rho_0 \phi^* + \frac{1}{2}A (\phi^*)^2 \right) + \frac{1}{2} A (\delta\phi)^2
$$
The term in the brackets is the **Hartree Energy shift**:
$$
\Delta S_{\text{Hartree}} = \frac{\rho_0^2}{A} - \frac{\rho_0^2}{2A} = \frac{\rho_0^2}{2A}
$$

### Integration
The partition function becomes:
$$
Z = e^{-(S_0 + \Delta S_{\text{Hartree}})} \int d(\delta\phi) \, e^{-\frac{1}{2} A (\delta\phi)^2} = e^{-(S_0 + \frac{\rho_0^2}{2A})} \sqrt{\frac{2\pi}{A}}
$$

### The Free Energy Formula
Using $F = -\ln Z$:

$$
F_{\text{RPA}} = S_0 + \underbrace{\frac{\rho_0^2}{2(\frac{\beta}{Cg} + \Pi)}}_{\text{Hartree (Energy Cost)}} + \underbrace{\frac{1}{2}\ln\left( \frac{\beta}{Cg} + \Pi \right)}_{\text{Correlation (Entropy)}}
$$

* **Hartree Term:** Represents the energy cost of the static mean field induced by the background density. It is positive for repulsive interactions.
* **Correlation Term:** Represents the entropy of Gaussian fluctuations (screening). It lowers the free energy.

### 3. Numerical Visualization

The following code calculates the RPA Free Energy components as a function of interaction strength $g$.


```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Parameters ---
beta = 5.0
xi = 0.5   # Energy level
# Vary Interaction g (assuming repulsive Coulomb-like, so g > 0)
# Note: In formula beta/(Cg), large g means small stiffness term.
g_vals = np.linspace(0.1, 5.0, 100)
C = 1.0

# --- 2. Calculation Functions ---

def fermi(E, beta):
    return 1 / (1 + np.exp(beta * E))

def calculate_rpa_free_energy(g_array, beta, xi):
    F_total = []
    F_hartree = []
    F_corr = []
    
    # 0. Basic Quantities (Independent of g)
    n_F = fermi(xi, beta)
    rho_0 = beta * n_F  # The coefficient of linear term (-i phi rho0) derived earlier
    
    # S_0 (Free gas free energy)
    # F0 = -ln(1 + e^-beta*xi)
    S_0 = -np.log(1 + np.exp(-beta * xi))
    
    # Polarization Pi = beta * n * (1-n)
    Pi = beta * n_F * (1 - n_F)
    
    for g in g_array:
        # Stiffness A = beta/(Cg) + Pi
        A = beta / (C * g) + Pi
        
        # 1. Hartree Term (from completing the square of linear term)
        # E_hartree = rho0^2 / (2A)
        f_hartree = (rho_0**2) / (2 * A)
        
        # 2. Correlation Term (Fluctuation Log)
        # E_corr = 0.5 * ln(A) - const
        f_corr = 0.5 * np.log(A) - 0.5 * np.log(2 * np.pi)
        
        F_hartree.append(f_hartree)
        F_corr.append(f_corr)
        F_total.append(S_0 + f_hartree + f_corr)
        
    return np.array(F_total), np.array(F_hartree), np.array(F_corr), S_0

# --- 3. Compute ---
F_tot, F_h, F_c, S0_val = calculate_rpa_free_energy(g_vals, beta, xi)

# --- 4. Plotting ---
plt.figure(figsize=(10, 6))

# Plot normalized changes
plt.plot(g_vals, F_tot, label='Total RPA Free Energy', color='black', linewidth=2)
plt.plot(g_vals, F_h + S0_val, label='Mean Field (S0 + Hartree)', color='blue', linestyle='--')
plt.plot(g_vals, F_c + S0_val, label='Fluctuation Only (S0 + Corr)', color='red', linestyle='-.')
plt.axhline(S0_val, color='gray', linestyle=':', label='Non-interacting Limit ($S_0$)')

plt.title(f'RPA Free Energy Components vs Interaction Strength $g$\n($\\beta={beta}, \\xi={xi}$)')
plt.xlabel('Interaction Strength $g$')
plt.ylabel('Free Energy $F$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Insight Text
plt.text(g_vals[-1]*0.6, F_tot[-1], "Hartree cost dominates\n(Repulsion increases Energy)", 
         ha='center', va='bottom', color='black')

plt.show()
```

### Exchange Channel (Magnetic / Fock)



This channel focuses on the scattering of a particle from state $p_1$ to $p_2$ and vice versa, often related to magnetic instabilities.

#### Operator Rearrangement
We rewrite the interaction to pair $\bar{\psi}_{p_1}$ with $\psi_{p_2}$.
Starting from $\bar{\psi}_{p_1} \bar{\psi}_{p_2} \psi_{p_2} \psi_{p_1}$:
1.  Swap $\bar{\psi}_{p_2}$ and $\psi_{p_2} \to - \bar{\psi}_{p_1} \psi_{p_2} \bar{\psi}_{p_2} \psi_{p_1}$ (plus a constant).
2.  Group terms as transition operators:

$$
S_{\text{int}}^{\text{exch}} = - C g_{\text{eff}} (\bar{\psi}_{p_1} \psi_{p_2}) (\bar{\psi}_{p_2} \psi_{p_1})
$$

#### HS Transformation
We introduce a complex auxiliary bosonic field $M(\tau)$ (and its conjugate $\bar{M}$) to decouple the term.
* **Field Interpretation:** $M$ represents a "hybridization" or spin-flip field derived from the Fock exchange potential.

$$
\exp\left( -S_{\text{int}}^{\text{exch}} \right) = \int \mathcal{D}M \mathcal{D}\bar{M} \exp \left[ - \int d\tau \left( \frac{|M|^2}{C g_{\text{eff}}} - M \bar{\psi}_{p_2} \psi_{p_1} - \bar{M} \bar{\psi}_{p_1} \psi_{p_2} \right) \right]
$$

#### Fermionic Path Integral
The fermionic action now mixes $p_1$ and $p_2$. We write it in vector form $\Psi = (\psi_{p_1}, \psi_{p_2})^T$:

$$
S_{\text{fermion}} = \sum_{\omega_n} \begin{pmatrix} \bar{\psi}_{p_1} & \bar{\psi}_{p_2} \end{pmatrix}
\underbrace{
\begin{pmatrix}
-i\omega_n + \xi_{p_1} & -\bar{M} \\
-M & -i\omega_n + \xi_{p_2}
\end{pmatrix}
}_{\mathcal{G}^{-1}_{\text{exch}}}
\begin{pmatrix}
\psi_{p_1} \\
\psi_{p_2}
\end{pmatrix}
$$

#### Effective Action
We calculate the determinant of the $2 \times 2$ matrix:
$$
\det(\mathcal{G}^{-1}_{\text{exch}}) = (-i\omega_n + \xi_{p_1})(-i\omega_n + \xi_{p_2}) - |M|^2
$$

Integrating out the fermions yields:

$$
S_{\text{eff}}[M] = \frac{\beta |M|^2}{C g_{\text{eff}}} - \sum_n \ln \left[ (-i\omega_n + \xi_{p_1})(-i\omega_n + \xi_{p_2}) - |M|^2 \right]
$$

---

#### Expansion in the Exchange Channel: Stoner Instability

In this section, we derive the effective action for the Exchange (Magnetic) channel by expanding the logarithm term up to the second order in the auxiliary field $M$. This provides the microscopic derivation of the Stoner criterion for ferromagnetism.

** 1. The Effective Action **

We start with the effective action derived via the Hubbard-Stratonovich transformation in the exchange channel:

$$
S_{\text{eff}}[M] = \frac{\beta |M|^2}{C g_{\text{eff}}} - \sum_n \ln \left[ (-i\omega_n + \xi_{1})(-i\omega_n + \xi_{2}) - |M|^2 \right]
$$

Here, $\xi_1$ and $\xi_2$ are the energies of the two states involved in the scattering, and $M$ represents the spin-flip or exchange field.

** 2. Taylor Expansion (Weak Field Limit) **

We aim to expand the action around the paramagnetic state ($M=0$). Let $A_n = (-i\omega_n + \xi_1)(-i\omega_n + \xi_2)$. The logarithmic term can be rewritten as:

$$
\ln(A_n - |M|^2) = \ln \left[ A_n \left( 1 - \frac{|M|^2}{A_n} \right) \right] = \ln(A_n) + \ln \left( 1 - \frac{|M|^2}{A_n} \right)
$$

Using the Taylor series $\ln(1-x) \approx -x - x^2/2 - \dots$, and keeping only the leading order term (which is quadratic in $M$):

$$
\ln \left( 1 - \frac{|M|^2}{A_n} \right) \approx - \frac{|M|^2}{(-i\omega_n + \xi_1)(-i\omega_n + \xi_2)}
$$

Substituting this back into the action (note the minus sign in front of the sum):

$$
S_{\text{eff}}[M] \approx S_0 + \frac{\beta |M|^2}{C g_{\text{eff}}} + |M|^2 \sum_n \frac{1}{(-i\omega_n + \xi_1)(-i\omega_n + \xi_2)}
$$

** 3. The Susceptibility $\chi_0$ **
The summation term corresponds to the static, non-interacting susceptibility (Lindhard function at $q=0, \omega=0$). We evaluate it using partial fraction decomposition:

$$
\frac{1}{(-i\omega_n + \xi_1)(-i\omega_n + \xi_2)} = \frac{1}{\xi_1 - \xi_2} \left( \frac{1}{-i\omega_n + \xi_2} - \frac{1}{-i\omega_n + \xi_1} \right)
$$

Summing over Matsubara frequencies $\omega_n$, we use the standard identity $T \sum_n \frac{1}{-i\omega_n + \xi} = n_F(\xi)$:

$$
\chi_0 \equiv - \frac{1}{\beta} \sum_n \frac{1}{A_n} = \frac{n_F(\xi_1) - n_F(\xi_2)}{\xi_1 - \xi_2}
$$

*(Note: The sign convention for $\chi_0$ implies $\chi_0 > 0$ for stable systems).*

** 4. Resulting Ginzburg-Landau Action **

Combining the terms, the effective action near $M=0$ is:

$$
S_{\text{eff}}^{(2)}[M] = S_0 + \beta |M|^2 \left( \frac{1}{C g_{\text{eff}}} - \chi_0 \right)
$$

Physical Interpretation

* **Coefficient > 0**: The Gaussian potential is positive. $M=0$ is the stable minimum. The system is paramagnetic.
* **Coefficient < 0**: The potential is unstable at $M=0$. $M$ will grow spontaneously. This is the **Stoner Criterion** for Ferromagnetism:
    $$C g_{\text{eff}} \chi_0 > 1$$

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Parameters ---
beta = 10.0
Cg = 2.0       # Effective coupling strength (Positive implies stable M=0 unless strong)
xi_1 = -0.5    # State 1
xi_2 = 0.5     # State 2 (Separated by 1.0)

# --- 2. Physics Functions ---

def fermi(E, beta):
    return 1 / (1 + np.exp(beta * E))

def action_exact_exchange(M, beta, Cg, xi1, xi2):
    # 1. Gaussian Term
    S_gauss = beta * (M**2) / Cg
    
    # 2. Fermionic Determinant Term (via Eigenvalues)
    # The eigenvalues of the 2x2 matrix with off-diagonal M
    avg_xi = (xi1 + xi2) / 2
    delta_xi = (xi1 - xi2) / 2
    
    # Eigenenergies E+ and E-
    root_term = np.sqrt(delta_xi**2 + M**2)
    E_plus = avg_xi + root_term
    E_minus = avg_xi - root_term
    
    # Sum ln(-iwn + E) -> ln(1 + e^-beta*E)
    S_fermion = -np.log(1 + np.exp(-beta * E_plus)) - np.log(1 + np.exp(-beta * E_minus))
    
    return S_gauss + S_fermion

def action_approx_exchange(M, beta, Cg, xi1, xi2):
    # 0th Order: M=0
    S_0 = action_exact_exchange(0, beta, Cg, xi1, xi2)
    
    # 2nd Order Coefficient: (1/g - chi_0)
    # Susceptibility chi_0 = (n1 - n2) / (xi1 - xi2)
    n1 = fermi(xi1, beta)
    n2 = fermi(xi2, beta)
    
    # Handle numerical stability if xi1 == xi2 (derivative limit)
    if abs(xi1 - xi2) < 1e-5:
        chi_0 = beta * n1 * (1 - n1)
    else:
        chi_0 = (n1 - n2) / (xi1 - xi2)
        
    coeff = beta * (1/Cg - chi_0)
    
    return S_0 + coeff * M**2

# --- 3. Plotting ---
M_vals = np.linspace(0, 2.0, 200) # Assuming M is real magnitude
S_exact = [action_exact_exchange(m, beta, Cg, xi_1, xi_2) for m in M_vals]
S_approx = [action_approx_exchange(m, beta, Cg, xi_1, xi_2) for m in M_vals]

plt.figure(figsize=(10, 6))
plt.plot(M_vals, S_exact, label='Exact Action', color='blue', linewidth=2)
plt.plot(M_vals, S_approx, label='Quadratic Approx (Stoner)', color='red', linestyle='--', linewidth=2)

plt.title(f'Exchange Channel: Exact vs Stoner Expansion\n($\\beta={beta}, Cg={Cg}, \\Delta\\xi={xi_2-xi_1}$)')
plt.xlabel('Magnetization Magnitude $|M|$')
plt.ylabel('Effective Action $S_{\\text{eff}}$')
plt.xlim(0, 2.0)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Insight annotation
plt.text(1.0, (min(S_exact)+max(S_exact))/2, 
         "Divergence implies\nnon-Gaussian effects\nat strong M", 
         fontsize=10, color='blue', ha='center',
         bbox=dict(facecolor='white', alpha=0.8))

plt.show()
```


### Cooper Channel

To decouple the quartic term, we define **Pairing Operators** (analogous to Cooper pairs):

$$
\begin{aligned}
\hat{\Delta} &= \psi_{p_2} \psi_{p_1} \\
\hat{\Delta}^\dagger &= \bar{\psi}_{p_1} \bar{\psi}_{p_2}
\end{aligned}
$$

The action is now written in a perfect square form suitable for the HS transformation:

$$
S_{\text{int}}^{\text{2-state}} = C g_{\text{eff}} \, \hat{\Delta}^\dagger \hat{\Delta}
$$

Using the complex Hubbard-Stratonovich transformation formula:

$$
e^{ - C g_{\text{eff}} \hat{\Delta}^\dagger \hat{\Delta} } \propto \int \mathcal{D}\Delta \mathcal{D}\bar{\Delta} \exp\left[ \frac{|\Delta|^2}{C g_{\text{eff}}} - \left( \bar{\Delta} \psi_{p_2} \psi_{p_1} + \Delta \bar{\psi}_{p_1} \bar{\psi}_{p_2} \right) \right]
$$

This successfully linearizes the fermion interaction by introducing an auxiliary bosonic field $\Delta$.


#### Total Action and Path Integral (Explicit Form)



Here, we construct the full path integral by explicitly writing out the free fermion terms and the Hubbard-Stratonovich interaction terms without using matrix notation.


The total action $S_{\text{total}}$ consists of three distinct parts:
$$
S_{\text{total}} = S_{\text{free}} + S_{\text{boson}} + S_{\text{coupling}}
$$

This describes the kinetic energy and chemical potential for the fermions in states $p_1$ and $p_2$. In imaginary time $\tau$:

$$
S_{\text{free}} = \int_0^\beta d\tau \left[ \bar{\psi}_{p_1} (\partial_\tau + \xi_{p_1}) \psi_{p_1} + \bar{\psi}_{p_2} (\partial_\tau + \xi_{p_2}) \psi_{p_2} \right]
$$

* $\xi_p = \epsilon_p - \mu$
* We sum over the two specific momentum states $p_1$ and $p_2$.

This term comes purely from the HS transformation and acts as the "mass term" or cost function for the auxiliary field $\Delta$:

$$
S_{\text{boson}} = \int_0^\beta d\tau \frac{|\Delta(\tau)|^2}{C g_{\text{eff}}}
$$

This term describes the physical process where fermions interact with the pairing field.
* $\bar{\Delta} \psi_{p_2} \psi_{p_1}$: The field $\bar{\Delta}$ destroys a pair of fermions (annihilates $p_1, p_2$).
* $\Delta \bar{\psi}_{p_1} \bar{\psi}_{p_2}$: The field $\Delta$ creates a pair of fermions (creates $p_1, p_2$).

$$
S_{\text{coupling}} = - \int_0^\beta d\tau \left( \bar{\Delta}(\tau) \psi_{p_2}(\tau) \psi_{p_1}(\tau) + \Delta(\tau) \bar{\psi}_{p_1}(\tau) \bar{\psi}_{p_2}(\tau) \right)
$$

---

#### The Final Explicit Path Integral


Combining all terms above, the partition function $Z$ is the integral over all fermion fields ($\psi, \bar{\psi}$) and the auxiliary boson field ($\Delta, \bar{\Delta}$):

$$
Z = \int \mathcal{D}\Delta \mathcal{D}\bar{\Delta} \mathcal{D}\bar{\psi}_{p_1} \mathcal{D}\psi_{p_1} \mathcal{D}\bar{\psi}_{p_2} \mathcal{D}\psi_{p_2} \exp\left( -S_{\text{total}} \right)
$$

With the full explicit action:

$$
\begin{aligned}
S_{\text{total}} = \int_0^\beta d\tau \bigg[ \quad
& \underbrace{\bar{\psi}_{p_1} (\partial_\tau + \xi_{p_1}) \psi_{p_1} + \bar{\psi}_{p_2} (\partial_\tau + \xi_{p_2}) \psi_{p_2}}_{\text{Normal Propagation}} \\
&+ \underbrace{\frac{\bar{\Delta} \Delta}{C g_{\text{eff}}}}_{\text{Auxiliary Field Cost}} \\
&- \underbrace{\left( \bar{\Delta} \psi_{p_2} \psi_{p_1} + \Delta \bar{\psi}_{p_1} \bar{\psi}_{p_2} \right)}_{\text{Pair Creation / Annihilation}}
\quad \bigg]
\end{aligned}
$$



#### Frequency Domain Representation


If we expand in Matsubara frequencies ($\partial_\tau \to -i\omega_n$), the action becomes a sum over $n$:

$$
\begin{aligned}
S_{\text{total}} = \frac{\beta |\Delta|^2}{C g_{\text{eff}}} + \sum_{n} \bigg[ \quad
& \bar{\psi}_{p_1, n} (-i\omega_n + \xi_{p_1}) \psi_{p_1, n} \\
+ \, & \bar{\psi}_{p_2, n} (-i\omega_n + \xi_{p_2}) \psi_{p_2, n} \\
- \, & \bar{\Delta} \psi_{p_2, -n} \psi_{p_1, n} - \Delta \bar{\psi}_{p_1, n} \bar{\psi}_{p_2, -n}
\quad \bigg]
\end{aligned}
$$
*(Assuming $\Delta$ is static/constant for simplicity here, otherwise $\Delta$ would carry frequency transfer).*



#### Sequential Integration of Fermion Fields


Instead of using matrix determinants, we integrate the fields one by one. We start by integrating out the field $\psi_{p_1}$ (and $\bar{\psi}_{p_1}$), treating $\psi_{p_2}$ as a background source.


#### Isolating the $p_1$ Terms


Let's look at the action for a single frequency $n$. We group all terms containing $\psi_{p_1}$ or $\bar{\psi}_{p_1}$:

$$
S_{\text{total}} = S_{p_1} + S_{p_2}
$$

Where $S_{p_1}$ contains all the dependence on particle 1:

$$
S_{p_1} = \bar{\psi}_{p_1} \underbrace{(-i\omega_n + \xi_{p_1})}_{A_1} \psi_{p_1} - \Delta \bar{\psi}_{p_1} \bar{\psi}_{p_2} - \bar{\Delta} \psi_{p_2} \psi_{p_1}
$$

And $S_{p_2}$ is the remaining term for particle 2:

$$
S_{p_2} = \bar{\psi}_{p_2} (-i\omega_{-n} + \xi_{p_2}) \psi_{p_2}
$$


#### Completing the Square for $\psi_{p_1}$


To integrate $\int \mathcal{D}\bar{\psi}_{p_1} \mathcal{D}\psi_{p_1} e^{-S_{p_1}}$, we need to rewrite the interaction terms as "sources" ($\eta$) coupled to the fields.

We need the form $\bar{\psi}_{p_1} \eta + \bar{\eta} \psi_{p_1}$.

1.  **Term 1:** $- \Delta \bar{\psi}_{p_1} \bar{\psi}_{p_2} = \bar{\psi}_{p_1} \underbrace{(-\Delta \bar{\psi}_{p_2})}_{\eta}$
2.  **Term 2:** $- \bar{\Delta} \psi_{p_2} \psi_{p_1} = + \bar{\Delta} \psi_{p_1} \psi_{p_2} = \underbrace{(\bar{\Delta} \psi_{p_2})}_{\bar{\eta}} \psi_{p_1}$
    *(Note: We swapped $\psi_{p_2}$ and $\psi_{p_1}$ causing a sign change, then grouped to the left).*

So the action for $p_1$ is:
$$
S_{p_1} = \bar{\psi}_{p_1} A_1 \psi_{p_1} + \bar{\psi}_{p_1} \eta + \bar{\eta} \psi_{p_1}
$$
With sources:
* $\eta = -\Delta \bar{\psi}_{p_2}$
* $\bar{\eta} = \bar{\Delta} \psi_{p_2}$


Using the Gaussian shift formula $\int e^{-(\bar{\psi} A \psi + \bar{\psi} \eta + \bar{\eta} \psi)} = \det(A) e^{\bar{\eta} A^{-1} \eta}$:

$$
\int \mathcal{D}\bar{\psi}_{p_1} \mathcal{D}\psi_{p_1} e^{-S_{p_1}} = \det(A_1) \exp\left( \bar{\eta} A_1^{-1} \eta \right)
$$


Now we calculate the shift term $\bar{\eta} A_1^{-1} \eta$ which will be added to the action of $p_2$.

$$
\begin{aligned}
\text{Shift} &= (\bar{\Delta} \psi_{p_2}) \frac{1}{-i\omega_n + \xi_{p_1}} (-\Delta \bar{\psi}_{p_2}) \\
&= - \frac{|\Delta|^2}{-i\omega_n + \xi_{p_1}} \psi_{p_2} \bar{\psi}_{p_2}
\end{aligned}
$$

We swap $\psi_{p_2}$ and $\bar{\psi}_{p_2}$ to match standard order (adding a minus sign):

$$
\text{Shift} = + \frac{|\Delta|^2}{-i\omega_n + \xi_{p_1}} \bar{\psi}_{p_2} \psi_{p_2}
$$


Now we construct the effective action for the remaining fermion $p_2$.
The total coefficient for $\bar{\psi}_{p_2} \psi_{p_2}$ is the original kinetic term plus the new shift term derived above.

$$
S_{\text{eff}}^{(p_2)} = \bar{\psi}_{p_2} \underbrace{\left[ (i\omega_n + \xi_{p_2}) + \frac{|\Delta|^2}{-i\omega_n + \xi_{p_1}} \right]}_{A_{\text{new}}} \psi_{p_2}
$$
*(Note: $-i\omega_{-n}$ became $i\omega_n$)*.

Now, integrate out $\psi_{p_2}$:
$$
\int \mathcal{D}\bar{\psi}_{p_2} \mathcal{D}\psi_{p_2} e^{-S_{\text{eff}}^{(p_2)}} = \det(A_{\text{new}})
$$


#### Final Result Check


The total partition function is the product of the two determinants we found:

$$
\begin{aligned}
Z &\propto \det(A_1) \times \det(A_{\text{new}}) \\
&= (-i\omega_n + \xi_{p_1}) \times \left[ (i\omega_n + \xi_{p_2}) + \frac{|\Delta|^2}{-i\omega_n + \xi_{p_1}} \right] \\
&= (-i\omega_n + \xi_{p_1})(i\omega_n + \xi_{p_2}) + |\Delta|^2
\end{aligned}
$$

This perfectly matches the determinant of the matrix derived in the previous method:
$$
\det \mathcal{M} = (i\omega_n - \xi_{p_1})(i\omega_n + \xi_{p_2}) - |\Delta|^2
$$
*(Note: Since $-i\omega_n + \xi_{p_1} = -(i\omega_n - \xi_{p_1})$, the product is $-(i\omega - \xi)(i\omega + \xi) + |\Delta|^2 = - [ (i\omega - \xi)(i\omega + \xi) - |\Delta|^2 ]$. The overall sign is usually absorbed or consistent).*

#### Final Effective Action for the Pairing Field $\Delta$

After integrating out all fermionic degrees of freedom ($\psi_{p_1}, \psi_{p_2}$), we arrive at a theory involving only the bosonic order parameter $\Delta$.


#### The Partition Function


The partition function is now a path integral over the complex scalar field $\Delta$:

$$
Z = \int \mathcal{D}\Delta \mathcal{D}\bar{\Delta} \exp\left( - S_{\text{eff}}[\Delta] \right)
$$


Combining the original quadratic cost term and the contribution from the fermion determinant derived in the previous steps:

$$
S_{\text{eff}}[\Delta] = S_{\text{condense}} + S_{\text{fluct}}
$$

This comes from the Hubbard-Stratonovich transformation itself:
$$
S_{\text{condense}} = \frac{\beta |\Delta|^2}{C g_{\text{eff}}}
$$

This term represents the free energy of the fermions moving in the background of the $\Delta$ field. Using the identity $\det(A) = \exp(\text{Tr} \ln A)$, or simply summing the log of the determinant eigenvalues we found:

$$
S_{\text{fluct}} = - \sum_{n=-\infty}^{\infty} \ln \left[ (\omega_n^2 + \xi_{p_1} \xi_{p_2} + |\Delta|^2) + i\omega_n(\xi_{p_2} - \xi_{p_1}) \right]
$$
*(Note: We used the result $\det = (i\omega - \xi_1)(i\omega + \xi_2) - |\Delta|^2$ and simplified signs inside the log).*


#### The Symmetric Case (Standard BCS Result)


If we assume the two states are degenerate or related by time-reversal symmetry such that $\xi_{p_1} = \xi_{p_2} = \xi$:

$$
S_{\text{eff}}[\Delta] = \frac{\beta |\Delta|^2}{C g_{\text{eff}}} - \sum_{n} \ln \left[ \beta^2 (\omega_n^2 + \underbrace{\xi^2 + |\Delta|^2}_{E^2}) \right]
$$

* $E = \sqrt{\xi^2 + |\Delta|^2}$ is the **Quasiparticle Excitation Energy**.
* The term $\beta^2$ is usually a constant factor absorbed into the measure.



#### Evaluation of the Frequency Sum (Optional)



We can explicitly perform the Matsubara summation $T \sum_n \ln(\omega_n^2 + E^2)$ to get the thermodynamic potential $\Omega$:

$$
S_{\text{eff}}[\Delta] \approx \beta \Omega = \beta \left[ \frac{|\Delta|^2}{C g_{\text{eff}}} - 2 T \ln \left( 2 \cosh \frac{\beta E}{2} \right) \right]
$$

Or in the zero-temperature limit ($\beta \to \infty$):
$$
E_{\text{ground}} = \frac{|\Delta|^2}{C g_{\text{eff}}} - E
$$
*(Minimizing this gives the Gap Equation).*


#### Expansion in the Cooper Channel: Superconductivity

In this section, we derive the effective action for the pairing field $\Delta$ (the superconducting order parameter). We expand the fluctuation term $S_{\text{fluct}}$ around the normal state ($\Delta = 0$) to obtain the Ginzburg-Landau functional.

$$
S_{\text{eff}}[\Delta] = \frac{\beta |\Delta|^2}{C g_{\text{eff}}} - \sum_{n} \ln \left[ (\omega_n^2 + \xi_{1} \xi_{2}) + i\omega_n(\xi_{2} - \xi_{1}) + |\Delta|^2 \right]
$$

** 1. Taylor Expansion  **

Let $A_n$ be the term inside the logarithm when $\Delta=0$. This represents the product of the Green's functions in the normal state:
$$
A_n = (-i\omega_n + \xi_1)(i\omega_n + \xi_2)
$$

We rewrite the log term:
$$
\ln(A_n + |\Delta|^2) = \ln \left[ A_n \left( 1 + \frac{|\Delta|^2}{A_n} \right) \right] = \ln A_n + \ln \left( 1 + \frac{|\Delta|^2}{A_n} \right)
$$

Expanding for small order parameter $\Delta$:
$$
\ln \left( 1 + \frac{|\Delta|^2}{A_n} \right) \approx \frac{|\Delta|^2}{A_n}
$$

Substituting this back into $S_{\text{eff}}$:
$$
S_{\text{eff}}^{(2)} \approx S_0 + |\Delta|^2 \left( \frac{\beta}{C g_{\text{eff}}} - \sum_n \frac{1}{A_n} \right)
$$

** 2. The Pair Susceptibility ($\chi_{\text{pair}}$) **

The coefficient of the quadratic term contains the **Pair Susceptibility**:
$$
\chi_{\text{pair}} = \sum_n \frac{1}{A_n} = \sum_n \frac{1}{(-i\omega_n + \xi_1)(i\omega_n + \xi_2)}
$$

To evaluate this sum, we rewrite the denominator. Note that $i\omega_n + \xi_2 = -(-i\omega_n - \xi_2)$.
$$
\frac{1}{A_n} = - \frac{1}{(-i\omega_n + \xi_1)(-i\omega_n - \xi_2)}
$$
Using partial fractions:
$$
\frac{1}{(-i\omega_n + \xi_1)(-i\omega_n - \xi_2)} = \frac{1}{\xi_1 + \xi_2} \left( \frac{1}{-i\omega_n - \xi_2} - \frac{1}{-i\omega_n + \xi_1} \right)
$$

Performing the Matsubara sum ($\sum \frac{1}{-i\omega_n + E} = \beta n_F(E)$):
$$
\sum_n \frac{1}{A_n} = - \frac{\beta}{\xi_1 + \xi_2} [ n_F(-\xi_2) - n_F(\xi_1) ]
$$
Using the identity $n_F(-x) = 1 - n_F(x)$:
$$
\chi_{\text{pair}} = \beta \frac{1 - n_F(\xi_1) - n_F(\xi_2)}{\xi_1 + \xi_2}
$$

**Special Case: Cooper Pairing ($\xi_1 = \xi_2 = \xi$)**
If the two electrons have the same energy (time-reversed partners), we take the limit $\xi_2 \to \xi_1$:
$$
\chi_{\text{pair}} = \beta \frac{1 - 2n_F(\xi)}{2\xi} = \frac{\beta}{2\xi} \tanh\left(\frac{\beta \xi}{2}\right)
$$
This is the famous **Cooper Logarithm** term.

** 3. Ginzburg-Landau Functional **

The effective action near $T_c$ becomes:
$$
S_{\text{GL}}[\Delta] = S_0 + \beta |\Delta|^2 \left( \frac{1}{C g_{\text{eff}}} - \frac{\tanh(\frac{\beta \xi}{2})}{2\xi} \right)
$$
Setting the term in the brackets to zero yields the **BCS Gap Equation**.

** 4. Numerical Comparison**

The following code compares the full effective action (using quasiparticle energies $E = \sqrt{\xi^2+\Delta^2}$) with the quadratic Ginzburg-Landau approximation.
 

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Parameters ---
beta = 20.0     # Low temperature (High Beta) to see the minimum clearly
Cg = 0.5        # Attractive coupling strength (positive convention here for |Delta|^2/g)
xi = 0.2        # Energy of the state near Fermi surface

# --- 2. Physics Functions ---

def action_exact_cooper(Delta, beta, Cg, xi):
    # 1. Condensation Energy cost
    S_cond = beta * (Delta**2) / Cg
    
    # 2. Fermionic Part
    # The eigenvalues of the Nambu matrix for Cooper pairs are E = +/- sqrt(xi^2 + Delta^2)
    E_qp = np.sqrt(xi**2 + Delta**2)
    
    # The term sum_n ln(...) converts to the free energy of these quasiparticles
    # F_fermion = - sum ln(1 + e^-beta*E) - sum ln(1 + e^beta*E) ... 
    # Simplified for particle-hole symmetry: - 2 * ln(2 * cosh(beta * E / 2))
    # But let's stick to the standard form relative to normal state:
    # S_fluct = - 2 * [ ln(1 + e^-beta*E) - ln(1 + e^-beta*xi) ] (shifted so S(0)=0 for comparison)
    
    term_super = -2 * np.log(1 + np.exp(-beta * E_qp))
    term_normal = -2 * np.log(1 + np.exp(-beta * xi)) # To normalize S(0) baseline
    
    # We also need to account for the energy shift -beta * (E - xi) if using grand potential
    # Let's use the explicit Log determinant sum formula (approximated by the integral result)
    # Ideally: S = Delta^2/g - 2 * sum_k [ log(cosh(beta Ek/2)) ]
    
    S_fermion = -2 * np.log( np.cosh(beta * E_qp / 2) / np.cosh(beta * xi / 2) )
    
    return S_cond + S_fermion

def action_approx_cooper(Delta, beta, Cg, xi):
    # Quadratic Ginzburg-Landau: a * |Delta|^2
    # Coeff a = beta * (1/g - tanh(beta*xi/2)/(2*xi))
    
    tanh_term = np.tanh(beta * xi / 2)
    chi_pair = tanh_term / (2 * xi)
    
    coeff = beta * (1/Cg - chi_pair)
    
    return coeff * (Delta**2)

# --- 3. Plotting ---
Delta_vals = np.linspace(0, 0.6, 200)
S_ex = [action_exact_cooper(d, beta, Cg, xi) for d in Delta_vals]
S_ap = [action_approx_cooper(d, beta, Cg, xi) for d in Delta_vals]

plt.figure(figsize=(10, 6))
plt.plot(Delta_vals, S_ex, label='Full BCS Action', color='green', linewidth=2)
plt.plot(Delta_vals, S_ap, label='Ginzburg-Landau (Quadratic)', color='orange', linestyle='--', linewidth=2)

# Find minimum of exact action
min_idx = np.argmin(S_ex)
min_Delta = Delta_vals[min_idx]
plt.scatter([min_Delta], [S_ex[min_idx]], color='green', zorder=5, label=f'BCS Gap $\Delta_0={min_Delta:.2f}$')

plt.title(f'Cooper Channel: BCS vs Ginzburg-Landau\n($\\beta={beta}, Cg={Cg}, \\xi={xi}$)')
plt.xlabel('Order Parameter $|\Delta|$')
plt.ylabel('Effective Action $S_{\\text{eff}}$')
plt.axhline(0, color='gray', linestyle=':', alpha=0.5)

# Interpretations
if S_ap[10] < 0:
    status = "Unstable (Superconducting)"
else:
    status = "Stable (Normal State)"

plt.text(0.4, 0.5 * max(S_ex), f"State: {status}", fontsize=12, color='green')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
