# Condensed Matter Field Theory: From Quartic Interactions to a Unified Polarization Function
**Lecture Notes on the Microscopic Effective Actions for CDW, SDW, and SC**



In these notes, we organize CDW, SDW, and superconducting instabilities within one common field-theoretic framework. The main idea is simple: start from quartic fermion interactions, introduce auxiliary fields, integrate out the fermions, and read off the collective response from the resulting effective action.



## The Electron Gas Model and the "Curse" of Quartic Terms



We begin with the action of an interacting electron system, written as a free part plus an interaction part:
$$S = \int_0^\beta d\tau \sum_k \bar{c}_k (\partial_\tau + \epsilon_k) c_k + S_{int}$$
The difficulty is that $S_{int}$ contains quartic Grassmann products of the form $\bar{c}\bar{c}cc$. As a result, the path integral $Z = \int \mathcal{D}\bar{c}\mathcal{D}c e^{-S}$ is no longer exactly solvable in closed form.



## The Hubbard-Stratonovich Transformation



To decouple the quartic term, we introduce a macroscopic **auxiliary bosonic field $\phi$** through a Gaussian integral identity. Different instabilities correspond to different decoupling channels:

* **CDW (charge channel):** Introduce a scalar field $\phi(q)$ coupled as $\phi \bar{c}_{k+q} c_k$ (particle-hole channel, particle-number conserving).
* **SDW (spin channel):** Introduce a vector field $\vec{m}(q)$ coupled as $\vec{m} \cdot (\bar{c}_{k+q} \vec{\sigma} c_k)$ (particle-hole channel with spin flip).
* **SC (Cooper channel):** Introduce a complex field $\Delta(q)$ coupled as $\Delta \bar{c}_{k\uparrow} \bar{c}_{-k\downarrow} + h.c.$ (particle-particle channel, breaking the U(1) gauge symmetry).

This is the step that replaces quartic fermion interactions with bilinear fermion couplings to collective fields.



## Nambu Spinors and the Quadratic Action



After the decoupling step, it is natural to group fermionic degrees of freedom into a generalized spinor $\Psi_k$. The action then takes the quadratic matrix form
$$S[\bar{\Psi}, \Psi, \phi] = \int d\tau \sum_k \bar{\Psi} [-\mathcal{G}^{-1}] \Psi + \frac{\phi^2}{V}$$
with
$$\mathcal{G}^{-1} = \mathcal{G}_0^{-1} - \hat{V} \phi$$
Here:
* $\mathcal{G}_0^{-1}$ is the bare Green's-function matrix in the normal state.
* $\hat{V}$ is the **vertex matrix**. For CDW it is $I$, for SDW it is $\vec{\sigma}$, and for SC it becomes an off-diagonal matrix such as $\sigma_x$ in Nambu space.



## Integrating Out the Fermions: The Exact Tr ln Action



Once the fermionic fields appear only quadratically, the Grassmann integral can be carried out exactly:
$$\int \mathcal{D}\bar{\Psi} \mathcal{D}\Psi \exp(-\bar{\Psi} M \Psi) = \det(M)$$
Using $\det(M) = \exp(\text{Tr} \ln M)$, we obtain an effective action depending only on the collective field:
$$S_{eff}[\phi] = \frac{\phi^2}{V} - \text{Tr} \ln (-\mathcal{G}^{-1})$$
This is the central exact result from which both mean-field theory and response theory follow.



## Taylor Expansion and the RPA Approximation



To study weak fluctuations, we expand the exact action around $\phi = 0$ and keep terms up to quadratic order:
$$\text{Tr}\ln(\mathcal{G}_0^{-1} - \hat{V}\phi) \approx \text{Tr}\ln(\mathcal{G}_0^{-1}) - \text{Tr}(\mathcal{G}_0 \hat{V} \phi) - \frac{1}{2} \text{Tr}(\mathcal{G}_0 \hat{V} \mathcal{G}_0 \hat{V} \phi^2)$$
The quadratic coefficient defines the **generalized susceptibility**:
$$S^{(2)}_{RPA} = \left( \frac{1}{V} - \chi(q, \omega) \right) \phi^2 \quad \implies \quad \chi(q, i\omega_m) = - \text{Tr}(\mathcal{G}_0 \hat{V} \mathcal{G}_0 \hat{V})$$
At this level, the instability is controlled by the sign change of the quadratic term.



## Matsubara Summation and the Unified Bubble Formula



Evaluating the trace gives
$$\chi(q, i\omega_m) = - \frac{1}{\beta} \sum_{k, i\nu_n} \text{tr}_{mat} [\mathcal{G}_0(k+q, i\nu_n+i\omega_m) \hat{V} \mathcal{G}_0(k, i\nu_n) \hat{V}]$$
We then convert the Matsubara sum into a contour integral using the pole structure of the Fermi function and the residue theorem. After analytic continuation $i\omega_m \to \omega + i\eta$, the result becomes

$$\chi(q, \omega) = - \sum_{k} \sum_{n, m} \underbrace{|\langle n, k+q | \hat{V} | m, k \rangle|^2}_{\text{coherence factor}} \frac{f(\xi_{k+q, n}) - f(\xi_{k, m})}{\xi_{k+q, n} - \xi_{k, m} - \omega - i\eta}$$

This is the unified bubble formula that appears across charge, spin, and pairing channels.



### Comparing the Three Major Cases



| Physical System | Auxiliary Field $\phi$ | Spinor Basis $\Psi_k$ | Vertex Matrix $\hat{V}$ | Divergent Channel | Physical Outcome |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **CDW** | Scalar (charge density) | $(c_{k\sigma}, c_{k+Q\sigma})^T$ | $\hat{I}$ (identity) | Particle-hole | Nesting-driven Lindhard response |
| **SDW** | Vector (spin moment) | $(c_{k\uparrow}, c_{k\downarrow})^T$ | $\vec{\sigma}$ (Pauli matrices) | Particle-hole | Spin-flip-assisted nesting divergence |
| **SC** | Complex scalar (Cooper pair) | $(c_{k\uparrow}, c_{-k\downarrow}^\dagger)^T$ (Nambu) | $\sigma_x$ or $\sigma_y$ | Particle-particle | Cooper instability with a characteristic $\tanh(\epsilon/2T)/\epsilon$ logarithmic divergence |

**Takeaway:** The formulas look different only because the spinor basis and vertex matrix change from one channel to another. Structurally, the response calculation is the same.



# Path Integral Derivation of the Charge Density Wave (CDW) Effective Action



In this section, we work through the CDW problem in a more concrete way. The goal is to see exactly how the microscopic dispersion enters the effective free energy.



## The Bare Fermionic Action



Consider a 2D electron gas on a lattice with bare dispersion $\epsilon_k$ and density-density interaction $V_q$. In imaginary time, the partition function is $Z = \int \mathcal{D}[\bar{c}, c] e^{-S[\bar{c}, c]}$, with bare action
$$S[\bar{c}, c] = \int_0^\beta d\tau \left[ \sum_{k, \sigma} \bar{c}_{k,\sigma} (\partial_\tau + \epsilon_k) c_{k,\sigma} + \frac{1}{2N} \sum_{q} V_q \rho_q \rho_{-q} \right]$$
Here, $\rho_q = \sum_{k,\sigma} \bar{c}_{k+q,\sigma} c_{k,\sigma}$ is the density operator at momentum $q$, and $\beta = 1/(k_B T)$.



## The Hubbard-Stratonovich Transformation



We decouple the quartic term $\rho_q \rho_{-q}$ by introducing a real auxiliary field $\phi_q(\tau)$:
$$S[\bar{c}, c, \phi] = \int_0^\beta d\tau \left[ \sum_{k,\sigma} \bar{c}_{k,\sigma} (\partial_\tau + \epsilon_k) c_{k,\sigma} + \frac{1}{2N}\sum_q \frac{\phi_q \phi_{-q}}{V_q} - \frac{1}{N}\sum_q \phi_q \rho_{-q} \right]$$
At this point the fermions are bilinear, while the collective CDW fluctuations are carried by $\phi_q$.



## The Static Mean-Field CDW Ansatz



To focus on a CDW instability at a nesting vector $Q$, we assume the field condenses into a static mode:
$$\phi_q(\tau) \to N \phi (\delta_{q, Q} + \delta_{q, -Q})$$
Substituting this ansatz into the action gives
$$S = \frac{\beta \phi^2}{V_Q} + \int_0^\beta d\tau \sum_{k,\sigma} \left[ \bar{c}_{k,\sigma} (\partial_\tau + \epsilon_k) c_{k,\sigma} - \phi (\bar{c}_{k+Q,\sigma} c_{k,\sigma} + \bar{c}_{k,\sigma} c_{k+Q,\sigma}) \right]$$
So the order parameter $\phi$ directly mixes the states at $k$ and $k+Q$.



## Matrix Formulation and Integrating Out Fermions



We now fold the Brillouin zone into the reduced Brillouin zone (RBZ) and introduce
$$\Psi_k^\dagger = (\bar{c}_k, \bar{c}_{k+Q})$$
Then the inverse Green's-function matrix is
$$\mathcal{G}^{-1}(k, i\omega_n) = \begin{pmatrix} -i\omega_n + \epsilon_k & -\phi \\ -\phi & -i\omega_n + \epsilon_{k+Q} \end{pmatrix}$$
Since the action is quadratic, the fermions can be integrated out exactly, leaving
$$S_{\text{eff}}(\phi) = \frac{\beta \phi^2}{V_Q} - \sum_{k \in \text{RBZ}, i\omega_n, \sigma} \ln \left[ \det \begin{pmatrix} -i\omega_n + \epsilon_k & -\phi \\ -\phi & -i\omega_n + \epsilon_{k+Q} \end{pmatrix} \right]$$



## Matsubara Summation and the Reconstructed Bands



The determinant factorizes as
$$\det \mathcal{G}^{-1} = (-i\omega_n + \epsilon_k)(-i\omega_n + \epsilon_{k+Q}) - \phi^2 = (-i\omega_n + E_k^+)(-i\omega_n + E_k^-)$$
where the reconstructed bands are
$$E_k^{\pm}(\phi) = \frac{\epsilon_k + \epsilon_{k+Q}}{2} \pm \sqrt{\left(\frac{\epsilon_k - \epsilon_{k+Q}}{2}\right)^2 + \phi^2}$$
Therefore,
$$S_{\text{eff}}(\phi) = \frac{\beta \phi^2}{V_Q} - 2 \sum_{k \in \text{RBZ}, i\omega_n} \left[ \ln(-i\omega_n + E_k^+) + \ln(-i\omega_n + E_k^-) \right]$$
In the zero-temperature limit, this leads to the ground-state free energy density
$$F_{\text{eff}}(\phi) = \frac{\phi^2}{V_Q} + \frac{2}{N} \sum_{k \in \text{RBZ}} \left[ E_k^-(\phi) - E_k^-(0) \right]$$
This is the nonlinear effective potential for the CDW order parameter.



## The RPA Limit (Small $\phi$ Expansion)



For small $\phi$, we expand
$$E_k^-(\phi) \approx \min(\epsilon_k, \epsilon_{k+Q}) - \frac{\phi^2}{|\epsilon_k - \epsilon_{k+Q}|}$$
Substituting this into the free energy yields the quadratic RPA form
$$F_{\text{RPA}}(\phi) = \left[ \frac{1}{V_Q} - \frac{1}{2}\chi_0(Q) \right] \phi^2$$
with static Lindhard susceptibility
$$\chi_0(Q) = - \frac{2}{N} \sum_{k \in \text{BZ}} \frac{f(\epsilon_k) - f(\epsilon_{k+Q})}{\epsilon_k - \epsilon_{k+Q}}$$
If the Fermi surface is perfectly nested, $\chi_0(Q)$ diverges logarithmically, and the ordered state is favored.



# Extracting Observable Physics from the Effective Action $S_{\text{eff}}[\phi]$



Having derived an effective action, we now ask what physical information it contains. In lecture-note language: once we know the free energy, what can we read off from it?



## The Absolute Ground State and the Self-Consistent Field Equation



The first step is to minimize the free energy:
$$\frac{\partial F_{\text{eff}}(\phi)}{\partial \phi} = 0$$
The solution $\phi_0(T)$ gives the physical order parameter at temperature $T$. In superconducting problems, this is precisely the origin of the familiar BCS self-consistent gap equation.



## Thermodynamic Properties



This section collects the most direct thermodynamic quantities obtained by substituting the equilibrium value $\phi_0$ back into the effective free energy.



### Condensation Energy



The condensation energy is the free energy difference between the normal and ordered states:
$$E_{\text{cond}} = F_{\text{eff}}(0, T) - F_{\text{eff}}(\phi_0, T)$$
It measures the energetic gain associated with symmetry breaking and therefore quantifies the stability of the ordered phase.



### Specific Heat and the Jump at $T_c$



The specific heat follows from the second temperature derivative of the equilibrium free energy:
$$C_v = -T \frac{\partial^2 F_{\text{eff}}(\phi_0, T)}{\partial T^2}$$
Near a continuous transition, this quantity typically displays the familiar jump at $T_c$.



## Collective Excitations and Dispersion Relations



To study collective modes, write
$$\phi(x, \tau) = \phi_0 + \delta \phi(x, \tau)$$
and expand the action to quadratic order in $\delta\phi$. The inverse propagator satisfies
$$\mathcal{D}^{-1}(q, \omega) = 0 \implies \text{solve for } \omega(q)$$
The resulting poles describe collective excitations such as the amplitude (Higgs) mode and, for complex order parameters, the phase (Goldstone or Anderson-Bogoliubov) mode.



## Response Functions to External Fields



If we couple the system to an external electromagnetic field, then integrating out the fermions generates terms of the form
$$S_{\text{eff}}[\phi_0, A] \supset \frac{1}{2} \sum_{q, \omega} K(q, \omega) A_q A_{-q}$$
The kernel $K(q, \omega)$ controls the electromagnetic response. In particular, it gives access to the superfluid density in the long-wavelength static limit and to the optical conductivity after analytic continuation.



# Mean-Field Hamiltonian for Coexisting SDW and $d_{x^2-y^2}$-Wave BCS Order



We now turn to a standard coexistence problem: SDW order together with $d_{x^2-y^2}$-wave superconductivity. The main task is to organize all coupled degrees of freedom into a single quadratic Hamiltonian.



## The Physical Model and Interaction Terms



Start from a Hamiltonian with one kinetic term and two interaction channels:
$$H = H_0 + H_U + H_V$$
where $H_0 = \sum_{k, \sigma} \epsilon_k c^\dagger_{k\sigma} c_{k\sigma}$ is the bare tight-binding dispersion.



## Hubbard-Stratonovich Decoupling



At the mean-field level, the quartic interactions are replaced by couplings to auxiliary order parameters. We discuss the SDW and BCS channels separately.



### SDW-Channel Decoupling (Repulsive $U$)



Assume antiferromagnetic order at $Q = (\pi, \pi)$ with magnetization along the $z$ axis, and introduce the staggered magnetization $M$. The decoupled mean-field Hamiltonian is
$$H_{SDW} = -M \sum_{k} \left( c^\dagger_{k+Q, \uparrow} c_{k, \uparrow} - c^\dagger_{k+Q, \downarrow} c_{k, \downarrow} + h.c. \right)$$
Because $S_z = \frac{1}{2}(n_\uparrow - n_\downarrow)$, spin-up and spin-down electrons couple to exchange fields with opposite signs.



### BCS-Channel Decoupling (Attractive $V$)



In the pairing channel, we introduce $\Delta_k$ and write
$$H_{BCS} = \sum_{k} \left( \Delta_k c^\dagger_{k, \uparrow} c^\dagger_{-k, \downarrow} + \Delta_k^* c_{-k, \downarrow} c_{k, \uparrow} \right)$$
For $d_{x^2-y^2}$-wave pairing,
$$\Delta_k = \Delta_0 (\cos k_x - \cos k_y)$$
and under translation by $Q=(\pi, \pi)$,
$$\Delta_{k+Q} = \Delta_0 (\cos(k_x+\pi) - \cos(k_y+\pi)) = -\Delta_k$$
The sign change will matter when we assemble the matrix Hamiltonian.



## The Generalized Four-Component Nambu Spinor



Because SDW order mixes $k$ with $k+Q$, while pairing mixes $k$ with $-k$, we need a basis that closes both momentum and particle-hole sectors. In the RBZ we define
$$\Psi_k = \begin{pmatrix} c_{k, \uparrow} \\ c_{k+Q, \uparrow} \\ c^\dagger_{-k, \downarrow} \\ c^\dagger_{-k-Q, \downarrow} \end{pmatrix}$$
This is the natural generalized Nambu spinor for the coexistence problem.



## The $4 \times 4$ Quadratic Hamiltonian Matrix



Using the fermionic anticommutation relation $c c^\dagger = 1 - c^\dagger c$ and dropping constants, the mean-field Hamiltonian becomes
$$H = \sum_{k \in RBZ} \Psi_k^\dagger \; \mathcal{H}_k \; \Psi_k$$
with
$$
\mathcal{H}_k = 
\begin{pmatrix}
\epsilon_k & M & \Delta_k & 0 \\
M & \epsilon_{k+Q} & 0 & \Delta_{k+Q} \\
\Delta_k^* & 0 & -\epsilon_k & M \\
0 & \Delta_{k+Q}^* & M & -\epsilon_{k+Q}
\end{pmatrix}
$$
This matrix simultaneously tracks band folding, particle-hole structure, and superconducting pairing.



## Physical Interpretation of the Matrix Blocks



It is useful to read the matrix block by block. The following subsections isolate the three most important structures.



### Particle Sector (Spin Up, Upper Left)



The particle block is
$$\mathcal{H}_{particle} = \begin{pmatrix} \epsilon_k & M \\ M & \epsilon_{k+Q} \end{pmatrix}$$
This is the SDW band-folding problem for spin-up electrons and shows how a magnetic gap opens in the particle sector.



### Hole Sector (Spin Down, Lower Right)



The hole block is
$$\mathcal{H}_{hole} = \begin{pmatrix} -\epsilon_k & M \\ M & -\epsilon_{k+Q} \end{pmatrix}$$
The kinetic term changes sign in the hole sector because $-\epsilon_{-k} = -\epsilon_k$. Although the SDW coupling for spin-down electrons starts with the opposite sign, the particle-hole rewrite introduces another minus sign, leaving the off-diagonal entry equal to $+M$.



### Superconducting Pairing Sector (Off-Diagonal Block)



The pairing block is
$$\mathcal{H}_{pair} = \begin{pmatrix} \Delta_k & 0 \\ 0 & \Delta_{k+Q} \end{pmatrix}$$
Using $\Delta_{k+Q} = -\Delta_k$, this becomes
$$\mathcal{H}_{pair}^{(d-wave)} = \begin{pmatrix} \Delta_k & 0 \\ 0 & -\Delta_k \end{pmatrix}$$
This sign structure is one of the key ingredients behind the distinction between nodal and antinodal behavior in the coexistence problem.



# Path Integral Derivation of How Spin Fluctuations Drive $d$-Wave Superconductivity



In these notes, the point is to see how a repulsive interaction can still generate superconductivity. The answer is that the momentum structure of the effective pairing interaction matters more than its sign at zero momentum.



## The Action and the Microscopic Interaction



Assume a general momentum-dependent interaction $V(k, k')$. In the Cooper channel, the interaction part of the action is
$$S_{int} = \int_0^\beta d\tau \sum_{k, k'} V(k, k') \bar{c}_{k\uparrow} \bar{c}_{-k\downarrow} c_{-k'\downarrow} c_{k'\uparrow}$$
where $\bar{c}, c$ are Grassmann variables.



## The Hubbard-Stratonovich (HS) Transformation



We decouple the quartic interaction by introducing complex auxiliary fields $\Delta_k(\tau)$ and $\bar{\Delta}_k(\tau)$. The important point is that the field carries a momentum label, so the gap is a genuine function of momentum space.

The decoupled form is
$$e^{-S_{int}} = \int \mathcal{D}[\Delta_k, \bar{\Delta}_k] \exp \left( \sum_{k, k'} \bar{\Delta}_k [V^{-1}]_{k, k'} \Delta_{k'} - \sum_k \left( \bar{\Delta}_k c_{-k\downarrow} c_{k\uparrow} + \Delta_k \bar{c}_{k\uparrow} \bar{c}_{-k\downarrow} \right) \right)$$
where $[V^{-1}]_{k, k'}$ is the inverse of the interaction matrix.



## Integrating Out the Fermions to Obtain the Effective Action



Introduce the Nambu spinor
$$\Psi_k = (c_{k\uparrow}, \bar{c}_{-k\downarrow})^T$$
Then the fermionic part of the action becomes
$$S_F = \sum_{k, \omega_n} \bar{\Psi}_k (-\mathcal{G}^{-1}_{k, \omega_n}) \Psi_k$$
with inverse Green's function
$$-\mathcal{G}^{-1}_{k, \omega_n} = \begin{pmatrix} -i\omega_n + \epsilon_k & \Delta_k \\ \bar{\Delta}_k & -i\omega_n - \epsilon_k \end{pmatrix}$$
Integrating out the fermions yields
$$S_{eff}[\Delta, \bar{\Delta}] = - \sum_{k, k'} \bar{\Delta}_k [V^{-1}]_{k, k'} \Delta_{k'} - \text{Tr} \ln (-\mathcal{G}^{-1}_{k, \omega_n})$$



## Minimizing the Action: Deriving the BCS Gap Equation



The saddle-point condition is
$$\frac{\delta S_{eff}}{\delta \bar{\Delta}_k} = 0$$
Differentiating the quadratic bosonic term gives $-\sum_{k'} [V^{-1}]_{k, k'} \Delta_{k'}$, while differentiating $\text{Tr} \ln$ extracts the anomalous Green's function $F_k = \langle c_{-k\downarrow} c_{k\uparrow} \rangle$.

After multiplying by $V(q, k)$ and summing over Matsubara frequencies, we obtain the self-consistent gap equation:
$$\Delta_k = -\sum_{k'} V(k, k') \frac{\Delta_{k'}}{2 E_{k'}} \tanh\left(\frac{\beta E_{k'}}{2}\right)$$
where $E_{k'} = \sqrt{\epsilon_{k'}^2 + |\Delta_{k'}|^2}$ is the Bogoliubov quasiparticle energy.



## How Do Spin Fluctuations Induce $d$-Wave Pairing?



In spin-fluctuation theories of the repulsive-$U$ Hubbard model, the effective interaction takes the schematic form
$$V(k, k') \approx U + \frac{3}{2} U^2 \chi_s(k - k')$$
Near half filling on the square lattice, $\chi_s(q)$ is strongly peaked at the antiferromagnetic wavevector $Q = (\pi, \pi)$. Therefore scattering processes with $k-k' \approx Q$ are strongly repulsive.

Keeping only that dominant channel gives
$$\Delta_k \approx - V(Q) \frac{\Delta_{k-Q}}{2 E_{k-Q}} \tanh\left(\frac{\beta E_{k-Q}}{2}\right)$$
Since $V(Q)$ is large and positive, consistency requires
$$\Delta_k \text{ and } \Delta_{k-(\pi, \pi)} \text{ must have opposite signs!}$$
The standard function with exactly this property is
$$\cos(k_x + \pi) - \cos(k_y + \pi) = -\cos(k_x) + \cos(k_y) = - (\cos k_x - \cos k_y)$$
Hence antiferromagnetic spin fluctuations naturally favor $d_{x^2-y^2}$-wave superconductivity.
