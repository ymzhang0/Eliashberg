# Path Integral Derivation of the Charge Density Wave (CDW) Effective Action

This document provides a rigorous, step-by-step derivation of the effective action for a Charge Density Wave (CDW) instability, illustrating how the bare tight-binding dispersion relation explicitly enters the macroscopic effective field theory.

## 1. The Bare Fermionic Action
Consider a 2D electron gas on a lattice with a bare tight-binding dispersion $\epsilon_k$ and a two-body density-density interaction $V_q$. In the imaginary-time ($\tau$) Matsubara formalism, the partition function is $Z = \int \mathcal{D}[\bar{c}, c] e^{-S[\bar{c}, c]}$, where the bare action is:
$$S[\bar{c}, c] = \int_0^\beta d\tau \left[ \sum_{k, \sigma} \bar{c}_{k,\sigma} (\partial_\tau + \epsilon_k) c_{k,\sigma} + \frac{1}{2N} \sum_{q} V_q \rho_q \rho_{-q} \right]$$
Here, $\rho_q = \sum_{k,\sigma} \bar{c}_{k+q,\sigma} c_{k,\sigma}$ is the electron density fluctuation operator at momentum $q$, and $\beta = 1/(k_B T)$.

## 2. The Hubbard-Stratonovich Transformation
To decouple the quartic interaction term ($\rho_q \rho_{-q}$), we introduce a real scalar auxiliary field $\phi_q(\tau)$ for each momentum $q$. This field physically represents the macroscopic charge density fluctuations. 

Using the Gaussian integral identity (shifting $\phi_q \to \phi_q - V_q \rho_q$), the action becomes strictly bilinear in the fermion fields:
$$S[\bar{c}, c, \phi] = \int_0^\beta d\tau \left[ \sum_{k,\sigma} \bar{c}_{k,\sigma} (\partial_\tau + \epsilon_k) c_{k,\sigma} + \frac{1}{2N}\sum_q \frac{\phi_q \phi_{-q}}{V_q} - \frac{1}{N}\sum_q \phi_q \rho_{-q} \right]$$

## 3. The Static Mean-Field CDW Ansatz
We are interested in a macroscopic CDW instability occurring at a specific nesting wavevector $Q$ (e.g., $Q = (\pi, \pi)$ for a half-filled square lattice). We assume the auxiliary field condenses into a static ($\omega_n = 0$), macroscopic, and uniform amplitude at this specific wavevector. 
We drop all dynamical fluctuations and set $\phi_q(\tau) \to N \phi (\delta_{q, Q} + \delta_{q, -Q})$. 

Substituting this into the action, the coupling term simplifies to:
$$S = \frac{\beta \phi^2}{V_Q} + \int_0^\beta d\tau \sum_{k,\sigma} \left[ \bar{c}_{k,\sigma} (\partial_\tau + \epsilon_k) c_{k,\sigma} - \phi (\bar{c}_{k+Q,\sigma} c_{k,\sigma} + \bar{c}_{k,\sigma} c_{k+Q,\sigma}) \right]$$

## 4. Matrix Formulation and Integrating Out Fermions
The macroscopic field $\phi$ explicitly mixes the fermion state at momentum $k$ with the state at $k+Q$. To diagonalize this, we fold the Brillouin Zone into the Reduced Brillouin Zone (RBZ) and introduce the two-component spinor $\Psi_k^\dagger = (\bar{c}_k, \bar{c}_{k+Q})$. 

Fourier transforming to Matsubara frequencies $i\omega_n$, the fermionic part of the action can be written as $\sum_{k \in \text{RBZ}, i\omega_n} \Psi_k^\dagger \mathcal{G}^{-1}(k, i\omega_n) \Psi_k$, where the inverse Green's function matrix is:
$$\mathcal{G}^{-1}(k, i\omega_n) = \begin{pmatrix} -i\omega_n + \epsilon_k & -\phi \\ -\phi & -i\omega_n + \epsilon_{k+Q} \end{pmatrix}$$

Because the action is now quadratic in the fermion fields, we can exactly integrate out $\bar{\Psi}, \Psi$ using the standard Gaussian Grassmann integral $\int \mathcal{D}[\bar{\Psi}, \Psi] e^{-\bar{\Psi}\mathcal{G}^{-1}\Psi} = \det(\mathcal{G}^{-1}) = \exp(\text{Tr} \ln \mathcal{G}^{-1})$. 

This leaves the exact effective action purely in terms of the auxiliary field $\phi$:
$$S_{\text{eff}}(\phi) = \frac{\beta \phi^2}{V_Q} - \sum_{k \in \text{RBZ}, i\omega_n, \sigma} \ln \left[ \det \begin{pmatrix} -i\omega_n + \epsilon_k & -\phi \\ -\phi & -i\omega_n + \epsilon_{k+Q} \end{pmatrix} \right]$$

## 5. Matsubara Summation and the Reconstructed Bands
We evaluate the determinant of the $2 \times 2$ matrix:
$$\det \mathcal{G}^{-1} = (-i\omega_n + \epsilon_k)(-i\omega_n + \epsilon_{k+Q}) - \phi^2 = (-i\omega_n + E_k^+)(-i\omega_n + E_k^-)$$
Where $E_k^\pm$ are the new, reconstructed energy bands separated by the CDW gap:
$$E_k^{\pm}(\phi) = \frac{\epsilon_k + \epsilon_{k+Q}}{2} \pm \sqrt{\left(\frac{\epsilon_k - \epsilon_{k+Q}}{2}\right)^2 + \phi^2}$$

Substituting the factored determinant back into the action:
$$S_{\text{eff}}(\phi) = \frac{\beta \phi^2}{V_Q} - 2 \sum_{k \in \text{RBZ}, i\omega_n} \left[ \ln(-i\omega_n + E_k^+) + \ln(-i\omega_n + E_k^-) \right]$$
(The factor of 2 accounts for the spin sum).

To find the ground state Free Energy density $F_{\text{eff}}(\phi) = S_{\text{eff}} / \beta$, we perform the Matsubara frequency summation. In the zero-temperature limit ($T \to 0, \beta \to \infty$), the sum $\frac{1}{\beta}\sum_{i\omega_n} \ln(-i\omega_n + E)$ simply returns the energy $E$ if the state is occupied ($E < 0$), and $0$ if it is empty. 

Assuming half-filling, the lower band $E_k^-$ is fully occupied and the upper band $E_k^+$ is completely empty. We subtract the background energy at $\phi=0$ to obtain the final, numerically computable expression:
$$F_{\text{eff}}(\phi) = \frac{\phi^2}{V_Q} + \frac{2}{N} \sum_{k \in \text{RBZ}} \left[ E_k^-(\phi) - E_k^-(0) \right]$$
This equation maps the tight-binding dispersion explicitly into the highly non-linear potential landscape of the CDW order parameter.

## 6. The RPA Limit (Small $\phi$ Expansion)
If we assume the fluctuations of the auxiliary field are infinitesimally small, we can Taylor expand the square root in $E_k^-(\phi)$ around $\phi = 0$:
$$E_k^-(\phi) \approx \min(\epsilon_k, \epsilon_{k+Q}) - \frac{\phi^2}{|\epsilon_k - \epsilon_{k+Q}|}$$

Inserting this into $F_{\text{eff}}(\phi)$ and summing over the full Brillouin Zone, the effective free energy truncates to a purely quadratic form (the RPA approximation):
$$F_{\text{RPA}}(\phi) = \left[ \frac{1}{V_Q} - \frac{1}{2}\chi_0(Q) \right] \phi^2$$
Where $\chi_0(Q)$ is the static Lindhard susceptibility (Polarization bubble):
$$\chi_0(Q) = - \frac{2}{N} \sum_{k \in \text{BZ}} \frac{f(\epsilon_k) - f(\epsilon_{k+Q})}{\epsilon_k - \epsilon_{k+Q}}$$
If the bare Fermi surface is perfectly nested ($\epsilon_{k+Q} = -\epsilon_k$), $\chi_0(Q)$ diverges logarithmically, causing the quadratic coefficient to become negative and driving the spontaneous symmetry-breaking phase transition.

# 从有效作用量 $S_{\text{eff}}[\phi]$ 提取可观测物理量

在量子多体物理中，辅助场 $\phi$ 的有效作用量（或自由能密度 $F_{\text{eff}}$）不仅是一个标量函数，更是一个**“生成泛函（Generating Functional）”**。一旦通过数值或解析手段获得了 $F_{\text{eff}}[\phi]$，我们可以通过对其求导或分析其涨落，严格导出系统几乎所有的宏观热力学与动力学可观测量。

---

## 1. 绝对基态与自洽场方程 (Gap Equation)

大自然倾向于使系统的自由能最小化。通过寻找有效作用量的极小值点，我们可以确定系统真实的序参量（如超导能隙、CDW 振幅）。

* **数学操作：** 对辅助场求一阶导数并令其为零。
  $$\frac{\partial F_{\text{eff}}(\phi)}{\partial \phi} = 0$$
* **物理意义：** 该方程的解 $\phi_0(T)$ 即为特定温度下的真实宏观序参量。这在微观理论中严丝合缝地对应着著名的 **BCS 自洽能隙方程**。通过扫描温度 $T$，可以绘制出序参量随温度演化的相图，并精确定位临界温度 $T_c$。

---

## 2. 热力学特性 (Thermodynamic Properties)

将求得的真实基态 $\phi_0$ 代回有效自由能函数中，即可获得系统的一系列核心热力学性质。

### 2.1 凝聚能 (Condensation Energy)
* **数学操作：** 计算正常态（$\phi=0$）与有序态（$\phi=\phi_0$）的自由能差值。
  $$E_{\text{cond}} = F_{\text{eff}}(0, T) - F_{\text{eff}}(\phi_0, T)$$
* **物理意义：** 衡量系统通过对称性破缺（形成库珀对或电荷密度波）所“赚取”的能量。它决定了该相变状态的绝对稳定性（如超导体的热力学临界磁场 $H_c$）。

### 2.2 比热容与跃变 (Specific Heat & Jump)
* **数学操作：** 对系统处于基态时的自由能求温度的二阶导数。
  $$C_v = -T \frac{\partial^2 F_{\text{eff}}(\phi_0, T)}{\partial T^2}$$
* **物理意义：** 在临界温度 $T_c$ 处，由于 $\phi_0$ 从零连续增长，比热容理论上会出现一个**明显的阶跃（Specific Heat Jump）**，这是实验上判定二级相变的绝对热力学标志。

---

## 3. 集体激发与色散关系 (Collective Modes)

序参量并非绝对静止，大自然允许辅助场在谷底 $\phi_0$ 附近发生时空涨落：$\phi(x, \tau) = \phi_0 + \delta \phi(x, \tau)$。通过将这种微小涨落代回泛函并展开至二次项，我们可以求出准粒子的色散关系。

* **数学操作：** 计算涨落场 $\delta\phi(q, \omega)$ 的逆传播子（Inverse Propagator）$\mathcal{D}^{-1}(q, \omega)$，并寻找其极点：
  $$\mathcal{D}^{-1}(q, \omega) = 0 \implies \text{求出 } \omega(q)$$
* **物理意义：** 解出的 $\omega(q)$ 即为集体激发模式的色散关系。
  * **振幅模式 (Amplitude Mode / Higgs Mode)：** 对应于 $\delta\phi$ 大小的剧烈震荡。在 $\phi_0$ 处的二阶导数 $\frac{\partial^2 F}{\partial \phi^2}$ 直接正比于该模式有效质量的平方。它具有巨大的能隙（通常为 $2\phi_0$）。
  * **相位模式 (Phase Mode / Goldstone Mode)：** 如果辅助场是复数（如超导态），相位的波动不改变能量（无能隙）。其色散关系通常表现为声学支 $\omega \propto q$（安德森-博戈留波夫模式）。

---

## 4. 外场响应函数 (External Field Responses)

有效作用量编码了系统对外部微扰的所有反应。若在推导 $S_{\text{eff}}$ 时通过极小耦合引入外加电磁场（矢量势 $\vec{A}$，即 $k \to k - eA$），积掉费米子后作用量中将出现对 $A$ 的二次响应项。

* **数学操作：** 提取有效作用量中关于矢量势 $A$ 的二次项系数核 $K(q, \omega)$。
  $$S_{\text{eff}}[\phi_0, A] \supset \frac{1}{2} \sum_{q, \omega} K(q, \omega) A_q A_{-q}$$
* **物理意义：** 响应核 $K(q, \omega)$ 是系统的电磁响应中枢。
  * **超流密度 (Superfluid Density)：** 取 $K(q, \omega)$ 在静场长波极限下（$\omega \to 0, q \to 0$）的值，可直接计算出超流体密度，这在宏观上对应于抗磁性（伦敦穿透深度 $\lambda_L$）。
  * **光学电导率 (Optical Conductivity)：** 对 $K(q, \omega)$ 进行解析延拓求其虚部，即可获得交流电导率 $\sigma(\omega)$。该物理量可直接与红外光学吸收光谱实验进行比对。
